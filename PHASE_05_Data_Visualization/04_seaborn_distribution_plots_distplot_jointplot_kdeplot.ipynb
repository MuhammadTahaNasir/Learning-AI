## 📝 Core Concepts & Explanations

### Visualizing Distributions with Seaborn
Seaborn makes statistical visualization easy. We can analyze the shapes of continuous variables using:
- `histplot`: Displays the distribution of a single numerical variable.
- `kdeplot`: Fits and plots a smooth kernel density estimate.
- `jointplot`: Creates a multi-panel plot showing the relationship between two variables along with marginal distributions.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

# KDE Plot
plt.figure(figsize=(8, 4))
sns.kdeplot(data=tips, x='total_bill', hue='time', fill=True, common_norm=False, palette='viridis')
plt.title('Total Bill KDE by Time of Day')
plt.show()

# Joint Plot
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex', color='purple')
plt.show()
```
