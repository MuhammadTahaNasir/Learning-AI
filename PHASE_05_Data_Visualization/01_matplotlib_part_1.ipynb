## 📝 Core Concepts & Explanations

### Matplotlib Fundamentals
Matplotlib is the foundation of data visualization in Python. It uses a hierarchy of objects:
- **Figure**: The overall window or page that holds the drawings.
- **Axes**: The actual plotting area where data is drawn (each figure can have multiple axes).

We can use the state-based interface (`plt.plot`) or the Object-Oriented (OO) interface (`fig, ax = plt.subplots()`). The OO interface is preferred for complex layouts.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Generate synthetic data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 2. Plot using Object-Oriented interface
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y1, label='Sin Wave', color='teal', linestyle='-', linewidth=2)
ax.scatter(x[::5], y2[::5], label='Cos Points', color='coral', marker='o')

ax.set_title('Sine Wave & Cosine Points')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.legend()
ax.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### Exercises
1. Plot a custom mathematical function $y = x^2 - 4x + 4$ for $x \in [-5, 5]$ as a dashed line. Add labels and a grid.

```python
x_custom = np.linspace(-5, 5, 100)
y_custom = x_custom**2 - 4*x_custom + 4
fig, ax = plt.subplots()
ax.plot(x_custom, y_custom, 'r--', label='quadratic')
ax.set_title('Quadratic Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.grid(True)
plt.show()
```
