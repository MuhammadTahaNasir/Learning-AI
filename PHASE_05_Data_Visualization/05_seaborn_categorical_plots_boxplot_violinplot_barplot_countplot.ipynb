## 📝 Core Concepts & Explanations

### Categorical Explorations
When analyzing data, we often need to examine distributions of numerical values across different categorical groups:
- `boxplot`: Displays five-number summaries and outlines outliers.
- `violinplot`: Combines a boxplot with a kernel density plot to reveal distribution density shape.
- `countplot`: Shows counts of observations in each categorical bin.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Boxplot
sns.boxplot(ax=axes[0], data=tips, x='day', y='total_bill', hue='smoker', palette='Set2')
axes[0].set_title('Boxplot of Total Bill by Day')

# Violinplot
sns.violinplot(ax=axes[1], data=tips, x='day', y='total_bill', hue='smoker', split=True, palette='muted')
axes[1].set_title('Violinplot of Total Bill by Day')

plt.tight_layout()
plt.show()
```
