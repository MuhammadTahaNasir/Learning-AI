## 📝 Core Concepts & Explanations

### Multi-variable Visual Inspections
Bivariate analysis explores correlations between pairs of columns, while multivariate analysis uses color hue, markers, and sizing to examine multiple variables simultaneously.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

# Bivariate Scatter plot with multi-dimensional dimensions
plt.figure(figsize=(8, 5))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', palette='deep')
plt.title('Bivariate Scatter Plot with Multi-variables')
plt.show()

# Bivariate Boxplot
plt.figure(figsize=(8, 4))
sns.boxplot(data=tips, x='time', y='total_bill', hue='sex', palette='coolwarm')
plt.title('Bivariate Comparison: Bill by Time & Sex')
plt.show()
```
