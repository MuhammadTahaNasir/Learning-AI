## 📝 Core Concepts & Explanations

### Univariate Analysis Checklist
Univariate analysis examines one feature at a time. It covers:
- **For Numeric Variables**: Statistics (mean, variance, quartiles) and plotting distributions via KDE or boxplot.
- **For Categorical Variables**: Frequency tables and count plots.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

# 1. Statistical Summary
print("=== Tip Column Statistics ===")
print(tips['tip'].describe())

# 2. Plotting distribution
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.histplot(ax=axes[0], data=tips, x='tip', kde=True, color='purple')
axes[0].set_title('Tip Distribution Histogram')

sns.boxplot(ax=axes[1], data=tips, x='tip', color='lightgreen')
axes[1].set_title('Tip Boxplot (Outlier Detection)')

plt.tight_layout()
plt.show()
```
