## 📝 Core Concepts & Explanations

### K-Medoids (PAM)
In K-Means, centroids are mathematical means (which might not be actual data points). In K-Medoids (Partitioning Around Medoids), the cluster centers are forced to be **actual data points** from the dataset.
- Because it uses absolute distances (L1 norm/Manhattan) and actual points, it is highly **robust to outliers** compared to K-Means.
- Computationally more expensive than K-Means.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import warnings
warnings.filterwarnings('ignore')
try:
    from sklearn_extra.cluster import KMedoids
    
    X, _ = make_blobs(n_samples=200, centers=3, random_state=42)
    # Add an extreme outlier
    X[0] = [20, 20]
    
    kmedoids = KMedoids(n_clusters=3, random_state=42).fit(X)
    
    plt.scatter(X[:, 0], X[:, 1], c=kmedoids.labels_, cmap='viridis')
    plt.scatter(kmedoids.cluster_centers_[:, 0], kmedoids.cluster_centers_[:, 1], c='red', marker='s', s=100, label='Medoids')
    plt.title('K-Medoids Clustering (Robust to Outliers)')
    plt.legend()
    plt.show()
except ImportError:
    print('scikit-learn-extra is required for KMedoids.')
```
