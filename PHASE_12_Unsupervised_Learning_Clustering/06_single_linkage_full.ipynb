## 📝 Core Concepts & Explanations

### Single Linkage
When merging two clusters in Agglomerative clustering, how do we measure the distance between the two clusters?
**Single Linkage** defines the distance between Cluster A and Cluster B as the distance between their **closest** points.
- **Pros:** Can handle non-elliptical shapes.
- **Cons:** Highly susceptible to **Chaining**. A single line of noise points can merge two entirely distinct clusters together.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import AgglomerativeClustering

X, _ = make_moons(n_samples=200, noise=0.05, random_state=42)

hc_single = AgglomerativeClustering(n_clusters=2, linkage='single')
labels = hc_single.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title('Single Linkage excelling at chaining (moons)')
plt.show()
```
