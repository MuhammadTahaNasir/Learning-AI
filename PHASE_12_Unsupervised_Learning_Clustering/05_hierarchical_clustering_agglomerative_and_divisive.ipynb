## 📝 Core Concepts & Explanations

### Hierarchical Clustering
A clustering algorithm that builds a hierarchy of clusters. It does not require specifying $K$ upfront.
1. **Agglomerative (Bottom-Up):** Start with $N$ points as $N$ clusters. Iteratively merge the closest pair until 1 massive cluster remains.
2. **Divisive (Top-Down):** Start with 1 massive cluster. Iteratively split it into smaller clusters until $N$ clusters exist.
