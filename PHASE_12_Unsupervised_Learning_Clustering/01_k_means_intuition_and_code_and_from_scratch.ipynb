## 📝 Core Concepts & Explanations

### K-Means Clustering Intuition
K-Means is an unsupervised learning algorithm that groups unlabeled data into $K$ distinct clusters.
**Algorithm Steps:**
1. Randomly initialize $K$ cluster centroids.
2. **Assignment Step:** Assign each data point to the closest centroid (using Euclidean distance).
3. **Update Step:** Recalculate each centroid as the mean of all data points assigned to it.
4. Repeat until centroids no longer move (convergence).

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# K-Means from Scratch (1 Iteration)
np.random.seed(42)
centroids = X[np.random.choice(X.shape[0], 4, replace=False)]

# 1. Assignment
distances = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)
labels = np.argmin(distances, axis=1)

# 2. Update
new_centroids = np.array([X[labels == i].mean(axis=0) for i in range(4)])

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.5)
plt.scatter(new_centroids[:, 0], new_centroids[:, 1], c='red', marker='x', s=200, label='Centroids')
plt.title('K-Means From Scratch: First Iteration')
plt.legend()
plt.show()
```
