## 📝 Core Concepts & Explanations

### K-Means in Practice
In Scikit-Learn, `KMeans` takes several important parameters:
- `n_clusters`: The number of clusters $K$.
- `n_init`: Number of times the algorithm runs with different centroid seeds. The best output (lowest WCSS) is kept.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import warnings
warnings.filterwarnings('ignore')

data = load_iris()
X = data.data[:, [2, 3]] # Using petal length and petal width for 2D visualization

kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)
labels = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('K-Means Clustering on Iris Dataset')
plt.legend()
plt.show()
```
