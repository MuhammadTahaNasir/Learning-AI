## 📝 Core Concepts & Explanations

### K-Means++ Initialization
To solve the poor initialization problem, **K-Means++** selects initial centroids strategically:
1. Choose the first centroid uniformly at random.
2. For each data point $x$, compute $D(x)$, the distance to the nearest existing centroid.
3. Choose the next centroid randomly with probability proportional to $D(x)^2$.
4. Repeat until $K$ centroids are chosen.
This ensures centroids are pushed as far apart from each other as possible before the algorithm begins.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# init='k-means++' is the default in scikit-learn
kmeans_plus_plus = KMeans(n_clusters=4, init='k-means++', n_init=10, random_state=42)
kmeans_plus_plus.fit(X)

print('K-Means++ Initialized Centroids:')
print(kmeans_plus_plus.cluster_centers_)
print(f'Iterations to converge: {kmeans_plus_plus.n_iter_}')
```
