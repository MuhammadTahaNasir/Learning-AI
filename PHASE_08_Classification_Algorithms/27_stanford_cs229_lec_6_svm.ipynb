## 📝 Core Concepts & Explanations

### Support Vector Classifier Optimization
Objective: Minimize $\frac{1}{2}\|w\|^2$ subject to $y^{(i)}(w^T x^{(i)} + b) \ge 1$.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
from sklearn.datasets import load_iris
from sklearn.svm import SVC

X, y = load_iris(return_X_y=True)
# Binary classification subset
X, y = X[y < 2], y[y < 2]

svm_model = SVC(kernel='linear')
svm_model.fit(X, y)
print('Support Vectors count:', len(svm_model.support_vectors_))
```
