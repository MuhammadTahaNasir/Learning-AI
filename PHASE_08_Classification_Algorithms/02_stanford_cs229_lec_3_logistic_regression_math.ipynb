## 📝 Core Concepts & Explanations

### Logistic Regression Math
- **Hypothesis:** $h_\theta(x) = g(\theta^T x) = \frac{1}{1 + e^{-\theta^T x}}$
- **Log-Likelihood:** $\ell(\theta) = \sum_{i=1}^m y^{(i)}\log h(x^{(i)}) + (1-y^{(i)})\log(1-h(x^{(i)}))$
- **Gradient:** $\nabla_\theta \ell(\theta) = X^T (y - h(x))$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
z = np.linspace(-10, 10, 100)
g = 1 / (1 + np.exp(-z))
dg = g * (1 - g)

plt.figure(figsize=(8, 4))
plt.plot(z, g, 'b-', label='Sigmoid g(z)')
plt.plot(z, dg, 'r--', label="Sigmoid Derivative g'(z)")
plt.axhline(0.5, color='gray', linestyle=':')
plt.axvline(0.0, color='gray', linestyle=':')
plt.title('Sigmoid Activation and its Derivative')
plt.xlabel('z')
plt.ylabel('Activation')
plt.legend()
plt.grid(True)
plt.show()
```

### Exercises
1. Calculate the log-likelihood for predictions $[0.99, 0.01]$ and labels $[1, 0]$.

```python
p_ex = np.array([0.99, 0.01])
y_ex = np.array([1, 0])
print('Log-Likelihood:', np.sum(y_ex * np.log(p_ex) + (1-y_ex) * np.log(1-p_ex)))
```
