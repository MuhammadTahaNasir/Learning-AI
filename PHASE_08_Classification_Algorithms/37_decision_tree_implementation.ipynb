## 📝 Core Concepts & Explanations

### Feature Importance
Impurity reduction importance: total Gini reduction contributed by each feature split.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.tree import DecisionTreeClassifier

wine = load_wine()
X = pd.DataFrame(wine.data, columns=wine.feature_names)
y = wine.target

clf = DecisionTreeClassifier(max_depth=3, random_state=42).fit(X, y)
imp = pd.Series(clf.feature_importances_, index=X.columns).sort_values(ascending=True)

plt.figure(figsize=(8, 5))
imp.plot(kind='barh', color='teal')
plt.title('Decision Tree Feature Importances (Wine Dataset)')
plt.xlabel('Gini Importance')
plt.tight_layout()
plt.show()
```
