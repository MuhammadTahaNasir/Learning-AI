### Mappings in SVM
Circular XOR classification separation using Radial Basis Function (RBF) mapping.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.svm import SVC

X, y = make_circles(n_samples=100, noise=0.1, factor=0.5, random_state=42)
clf_rbf = SVC(kernel='rbf', C=1.0).fit(X, y)

xx, yy = np.meshgrid(np.linspace(-1.5, 1.5, 100), np.linspace(-1.5, 1.5, 100))
Z = clf_rbf.decision_function(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.figure(figsize=(6, 5))
plt.contourf(xx, yy, Z, levels=20, cmap='RdYlBu', alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='RdYlBu', edgecolors='k')
plt.title('RBF Kernel SVM Decision Boundary')
plt.colorbar(label='Decision Score')
plt.show()

clf_linear = SVC(kernel='linear').fit(X, y)
print('Linear kernel accuracy:', clf_linear.score(X, y))
print('RBF kernel accuracy:', clf_rbf.score(X, y))
```
