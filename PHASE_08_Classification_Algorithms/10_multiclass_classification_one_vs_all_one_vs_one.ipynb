## 📝 Core Concepts & Explanations

### OvR and OvO
- **One-vs-Rest (OvR):** Fits $K$ classifiers. Fast but may suffer from size imbalances.
- **One-vs-One (OvO):** Fits $\frac{K(K-1)}{2}$ classifiers. Slower but robust.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
from sklearn.multiclass import OneVsRestClassifier, OneVsOneClassifier
from sklearn.svm import SVC
from sklearn.datasets import load_wine

wine = load_wine()
X, y = wine.data, wine.target

ovr = OneVsRestClassifier(SVC(kernel='linear')).fit(X, y)
ovo = OneVsOneClassifier(SVC(kernel='linear')).fit(X, y)
print('OvR score:', ovr.score(X, y))
print('OvO score:', ovo.score(X, y))
```
