## 📝 Core Concepts & Explanations

### Tree Generation Algorithms
- **ID3:** Relies on Information Gain. Only supports categorical splits.
- **C4.5:** Uses Gain Ratio to penalize multi-way branching bias.
- **CART:** Uses Gini Impurity. Restricts splits to binary choices.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris

iris = load_iris()
# CART algorithm is implemented in Scikit-Learn
clf = DecisionTreeClassifier(criterion='gini', random_state=42)
clf.fit(iris.data, iris.target)
print('CART Tree Score:', clf.score(iris.data, iris.target))
```
