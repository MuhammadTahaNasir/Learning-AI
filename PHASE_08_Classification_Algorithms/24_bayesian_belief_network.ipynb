## 📝 Core Concepts & Explanations

### Joint Factorization
$$P(X_1, X_2, X_3) = P(X_1)P(X_2|X_1)P(X_3|X_2)$$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
# Calculating joint probability of path in simple network
p_rain = 0.2
p_sprinkler_given_rain = 0.01
p_sprinkler_given_no_rain = 0.4

# P(Rain = True, Sprinkler = True)
p_joint = p_rain * p_sprinkler_given_rain
print('Joint Probability:', p_joint)
```

### Exercises
1. Compute P(Rain = False, Sprinkler = True).

```python
p_joint_no_rain = (1 - p_rain) * p_sprinkler_given_no_rain
print('Joint Probability (No Rain):', p_joint_no_rain)
```
