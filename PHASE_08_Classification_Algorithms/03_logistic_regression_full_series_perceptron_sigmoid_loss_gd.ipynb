## 📝 Core Concepts & Explanations

### Binary Cross Entropy (BCE) Loss
$$L(\theta) = -\frac{1}{m} \sum_{i=1}^m [y^{(i)}\log(\hat{y}^{(i)}) + (1-y^{(i)})\log(1-\hat{y}^{(i)})]$$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import matplotlib.pyplot as plt

p = np.linspace(0.01, 0.99, 100)
loss_y1 = -np.log(p)
loss_y0 = -np.log(1 - p)

plt.figure(figsize=(8, 4))
plt.plot(p, loss_y1, 'g-', label='BCE Loss for y=1: -log(p)')
plt.plot(p, loss_y0, 'r--', label='BCE Loss for y=0: -log(1-p)')
plt.xlabel('Predicted Probability p')
plt.ylabel('Loss Value')
plt.title('Binary Cross Entropy Loss Curves')
plt.legend()
plt.grid(True)
plt.show()
```

### Exercises
1. Compute the binary cross entropy loss for true label 0 and prediction 0.25.

```python
def bce_loss(y_true, y_pred):
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
print('BCE Loss:', bce_loss(0, 0.25))
```
