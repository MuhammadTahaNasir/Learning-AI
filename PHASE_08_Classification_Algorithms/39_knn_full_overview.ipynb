## 📝 Core Concepts & Explanations

### Distance calculation sensitivity
KNN is a lazy learner (no explicit training phase). It is highly sensitive to features magnitude differences.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

# Unscaled dataset where feature 1 has large magnitude
X = np.array([[1.0, 1000.0], [1.1, 1010.0], [5.0, 10.0], [5.1, 15.0]])
y = np.array([0, 0, 1, 1])

knn = KNeighborsClassifier(n_neighbors=1).fit(X, y)
# Predict query point close to class 1 features, but slightly shifted on feature 2
print('Prediction for [5.0, 100.0]:', knn.predict([[5.0, 100.0]])[0])
```
