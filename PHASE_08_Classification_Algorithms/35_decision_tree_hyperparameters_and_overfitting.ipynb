**Summary:** Regularizing decision trees to prevent overfitting.

---

## 📝 Core Concepts & Explanations

### Controlling Tree Expansion
- `max_depth`: Limits maximum vertical layers.
- `min_samples_split`: Samples needed to split a node.
- `min_samples_leaf`: Minimum samples allowed in a leaf.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

data = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.3, random_state=42)

clf_overfit = DecisionTreeClassifier(random_state=42)
clf_overfit.fit(X_train, y_train)

clf_reg = DecisionTreeClassifier(max_depth=3, min_samples_leaf=5, random_state=42)
clf_reg.fit(X_train, y_train)

print('Overfit test score:', clf_overfit.score(X_test, y_test))
print('Regularized test score:', clf_reg.score(X_test, y_test))
```
