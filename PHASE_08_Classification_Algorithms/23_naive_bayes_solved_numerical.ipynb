### Hand Numerical Example
Classify text given word occurrence counts and prior distribution probabilities.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

# Priors
prior_spam = 0.4
prior_ham = 0.6

# Word probabilities: [P(free|Spam), P(free|Ham)]
p_free = [0.8, 0.1]

# Score for document containing 'free'
score_spam = prior_spam * p_free[0]
score_ham = prior_ham * p_free[1]
print('Spam score:', score_spam)
print('Ham score:', score_ham)
```

### Exercises
1. Classify the document based on the calculated scores.

```python
print('Decision: SPAM' if score_spam > score_ham else 'Decision: HAM')
```
