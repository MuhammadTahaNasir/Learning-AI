## 📝 Core Concepts & Explanations

### Split Criteria
- **Entropy:** $H(S) = -\sum p_i \log_2 p_i$
- **Gini:** $G(S) = 1 - \sum p_i^2$
- **Information Gain:** $IG(S, A) = H(S) - \sum \frac{|S_v|}{|S|} H(S_v)$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

def entropy(probs):
    return -np.sum([p * np.log2(p) for p in probs if p > 0])

def gini(probs):
    return 1 - np.sum([p**2 for p in probs])

print('Entropy of [0.5, 0.5] uniform split:', entropy([0.5, 0.5]))
print('Gini Impurity of [0.5, 0.5] uniform split:', gini([0.5, 0.5]))
```
