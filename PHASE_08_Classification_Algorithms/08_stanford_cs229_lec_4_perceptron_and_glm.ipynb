## 📝 Core Concepts & Explanations

### GLM Formulation
1. $y|x; \theta \sim \text{ExponentialFamily}(\eta)$
2. $h_\theta(x) = E[y|x]$
3. $\eta = \theta^T x$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

def log_odds(p):
    return np.log(p / (1 - p))

print('Log-odds of 0.8 probability:', log_odds(0.8))
```
