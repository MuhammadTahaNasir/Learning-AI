## 📝 Core Concepts & Explanations

### Laplace Smoothing
$$P(x_j|y) = \frac{N_{jc} + \alpha}{N_c + \alpha D}$$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

# Count matrix for single feature: [counts class 0, counts class 1]
counts = np.array([0, 5])
n_0 = 10
n_1 = 12
vocab = 4

# Unsmoothed
p_raw_0 = counts[0] / n_0
# Smoothed (alpha=1)
p_smooth_0 = (counts[0] + 1) / (n_0 + vocab)
print(f'Raw: {p_raw_0}, Smoothed: {p_smooth_0}')
```
