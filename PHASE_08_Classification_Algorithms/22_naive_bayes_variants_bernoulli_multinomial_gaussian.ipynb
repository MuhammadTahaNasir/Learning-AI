## 📝 Core Concepts & Explanations

### NB Scikit-Learn Variants
- **GaussianNB:** Inputs are continuous normal values.
- **MultinomialNB:** Inputs are discrete word/feature occurrences.
- **BernoulliNB:** Inputs are binary indicators.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
from sklearn.naive_bayes import GaussianNB, BernoulliNB
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)
gnb = GaussianNB().fit(X, y)
bnb = BernoulliNB().fit(X, y)
print('Gaussian NB score:', gnb.score(X, y))
print('Bernoulli NB score:', bnb.score(X, y))
```
