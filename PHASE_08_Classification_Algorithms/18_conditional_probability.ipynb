## 📝 Core Concepts & Explanations

### Formula
$$P(A|B) = \frac{P(A \cap B)}{P(B)}$$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd

# Toy survey data
df = pd.DataFrame({
    'Purchased': [1, 0, 1, 1, 0, 1],
    'Gender': ['F', 'M', 'F', 'M', 'F', 'F']
})

# P(Purchased | Female)
female_subset = df[df['Gender'] == 'F']
p_cond = (female_subset['Purchased'] == 1).mean()
print('P(Purchased | Female) =', p_cond)
```
