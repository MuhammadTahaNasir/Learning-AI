## 📝 Core Concepts & Explanations

### Inductive Bias Examples
- **Linear Models (Logistic Regression):** Assume the class separation boundary is a linear hyperplane.
- **Decision Trees (CART):** Assume boundaries are orthogonal splits parallel to feature axes.
- **Nearest Neighbors (KNN):** Assume decision boundaries are determined locally by neighborhood clustering (Voronoi-like cells).
- **Support Vector Machines (Kernel SVM):** Assume boundaries can be mapped to high-dimensional spaces to find smooth, maximum-margin non-linear curves.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

X, y = make_moons(n_samples=150, noise=0.25, random_state=42)
models = {
    'Logistic Regression (Linear Bias)': LogisticRegression(),
    'Decision Tree (Axis-Aligned Bias)': DecisionTreeClassifier(max_depth=4, random_state=42),
    'KNN (K=3) (Local Instance Bias)': KNeighborsClassifier(n_neighbors=3),
    'SVM (RBF) (Radial Distance Bias)': SVC(kernel='rbf', C=1.0, random_state=42)
}

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.ravel()
xx, yy = np.meshgrid(np.linspace(-2, 3, 100), np.linspace(-1.5, 2, 100))
grid = np.c_[xx.ravel(), yy.ravel()]

for ax, (name, model) in zip(axes, models.items()):
    model.fit(X, y)
    Z = model.predict(grid).reshape(xx.shape)
    ax.contourf(xx, yy, Z, cmap='RdYlBu', alpha=0.6)
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap='RdYlBu', edgecolors='k')
    ax.set_title(name, fontsize=12)
    ax.grid(True, linestyle=':', alpha=0.5)

plt.suptitle('Classifier Inductive Bias Comparison (4 Classification Maps)', fontsize=16, y=0.98)
plt.tight_layout()
plt.show()

print('Inductive bias represents the set of assumptions a learner uses to predict outputs for unseen inputs.')
```
