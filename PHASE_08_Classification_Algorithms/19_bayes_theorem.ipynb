## 📝 Core Concepts & Explanations

### Updating Probabilities
$$P(A|B) = \frac{P(B|A)P(A)}{P(B)}$$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
# Calculate updated probability after positive medical test
prev = 0.01      # Prior P(Disease)
sens = 0.99      # Likelihood P(Pos | Disease)
spec = 0.95      # P(Neg | No Disease)

# Marginal P(Pos)
p_pos = sens * prev + (1 - spec) * (1 - prev)
# Posterior P(Disease | Pos)
post = sens * prev / p_pos
print(f'Posterior Probability: {post:.4f}')
```

### Exercises
1. Re-calculate the posterior probability if disease prevalence rises to 5%.

```python
prev_new = 0.05
p_pos_new = sens * prev_new + (1 - spec) * (1 - prev_new)
post_new = sens * prev_new / p_pos_new
print(f'New Posterior Probability: {post_new:.4f}')
```
