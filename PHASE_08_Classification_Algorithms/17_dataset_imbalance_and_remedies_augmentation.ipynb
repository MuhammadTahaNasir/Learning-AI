## 📝 Core Concepts & Explanations

### Remedying Imbalance
- Cost-sensitive learning: increasing misclassification penalty of minority class.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

X, y = make_classification(n_samples=200, n_features=4, weights=[0.9, 0.1], random_state=42)

lr_unbalanced = LogisticRegression().fit(X, y)
lr_balanced = LogisticRegression(class_weight='balanced').fit(X, y)

print('Unbalanced class predictions:', lr_unbalanced.predict(X).sum())
print('Balanced class predictions:', lr_balanced.predict(X).sum())
```
