## 📝 Core Concepts & Explanations

### Regularization Strength (C)
- $C = \frac{1}{\lambda}$: Smaller $C$ leads to stronger regularization.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression

data = load_breast_cancer()
X, y = data.data, data.target

model_l1 = LogisticRegression(penalty='l1', solver='liblinear', C=0.1, max_iter=10000, random_state=42)
model_l1.fit(X, y)
print('L1 non-zero coefficients:', np.sum(model_l1.coef_ != 0))
```

### Exercises
1. Fit a LogisticRegression model using L2 regularization and C=0.01. Compare training score with L1.

```python
model_l2 = LogisticRegression(penalty='l2', C=0.01, max_iter=10000, random_state=42).fit(X, y)
print('L2 score:', model_l2.score(X, y))
```
