## 📝 Core Concepts & Explanations

### GDA and Naive Bayes
- **GDA:** Assumes $x|y \sim N(\mu, \Sigma)$. Best for continuous variables.
- **Naive Bayes:** Assumes features are conditionally independent given label $y$.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
from sklearn.datasets import load_iris
from sklearn.naive_bayes import GaussianNB

iris = load_iris()
model = GaussianNB().fit(iris.data, iris.target)
print('Class priors:', model.class_prior_)
```
