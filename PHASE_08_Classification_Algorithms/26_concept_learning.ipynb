## 📝 Core Concepts & Explanations

### Hypothesis Space
- **Find-S:** Finds the most specific hypothesis consistent with positive training examples.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

# Instances [Outlook, Temp], label
data = [
    (['Sunny', 'Hot'], 1),
    (['Sunny', 'Warm'], 1),
    (['Rainy', 'Cold'], 0)
]

hypothesis = ['pi', 'pi'] # most specific

for attrs, label in data:
    if label == 1:
        for j in range(len(attrs)):
            if hypothesis[j] == 'pi':
                hypothesis[j] = attrs[j]
            elif hypothesis[j] != attrs[j]:
                hypothesis[j] = '?'
print('Final Hypothesis:', hypothesis)
```
