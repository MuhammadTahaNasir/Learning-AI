## 📝 Core Concepts & Explanations

### Tree Visualization Techniques
- **`plot_tree`:** Renders a graphical diagram of the tree splits using Matplotlib.
- **`export_text`:** Generates a text-based hierarchy of the decision rules, which is lightweight and fast.
- **Decision Boundaries:** Fitting on 2 features allows plotting 2D axis-aligned decision split maps.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text

iris = load_iris()
clf = DecisionTreeClassifier(max_depth=3, random_state=42)
clf.fit(iris.data, iris.target)

# 1. Graphical Plot
plt.figure(figsize=(10, 8))
plot_tree(clf, feature_names=list(iris.feature_names), class_names=list(iris.target_names), filled=True, rounded=True)
plt.title('Decision Tree structure on Iris dataset')
plt.show()

# 2. Text Representation
print('--- Decision Tree Rules (Text) ---')
print(export_text(clf, feature_names=list(iris.feature_names)))
```
