## 📝 Core Concepts & Explanations

### GD Update Calculation
Weight update: $\theta_j := \theta_j - \alpha (h_\theta(x) - y)x_j$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

X = np.array([1.0, 2.0]) # feature vector (including bias)
y = 1.0
theta = np.array([0.0, 0.5])
alpha = 0.1

# Sigmoid prediction
z = np.dot(theta, X)
h = 1 / (1 + np.exp(-z))

# Gradient step
theta_new = theta - alpha * (h - y) * X
print('New Theta:', theta_new)
```
