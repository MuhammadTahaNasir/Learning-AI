**Summary:** Comparison between regression and classification boundary limits.

---

## 📝 Core Concepts & Explanations

### Limits of Linear Regression for Classification
- **Predicting out of bounds:** Linear regression predicts values in $(-\infty, \infty)$, but probabilities must be in $[0, 1]$.
- **Sensitivity to outliers:** Adding extreme points shifts the decision boundary significantly, leading to misclassifications.
- **Non-constant variance:** Error terms violate homoscedasticity assumption in binary target scenarios.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression

# Generate toy binary dataset
X = np.array([1, 2, 3, 4, 10, 11, 12, 13]).reshape(-1, 1)
y = np.array([0, 0, 0, 0, 1, 1, 1, 1])

lin_reg = LinearRegression().fit(X, y)
log_reg = LogisticRegression().fit(X, y)

plt.scatter(X, y, color='black', label='Data')
plt.plot(X, lin_reg.predict(X), 'r--', label='Linear Regression')
plt.plot(X, log_reg.predict_proba(X)[:, 1], 'g-', label='Logistic Regression probability')
plt.axhline(0.5, color='gray', linestyle=':')
plt.legend()
plt.show()
```
