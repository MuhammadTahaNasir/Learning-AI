## 📝 Core Concepts & Explanations

### Optimal Classification Boundary
The decision boundary where class conditional probability densities intersect: $P(Y=0|x) = P(Y=1|x)$.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-5, 5, 200)
p_x_c0 = norm.pdf(x, loc=-1.0, scale=1.0) # class 0 density
p_x_c1 = norm.pdf(x, loc=1.0, scale=1.0)  # class 1 density

plt.plot(x, p_x_c0, 'r', label='Class 0')
plt.plot(x, p_x_c1, 'g', label='Class 1')
plt.axvline(0.0, color='black', linestyle='--', label='Bayes Boundary')
plt.legend()
plt.show()
```
