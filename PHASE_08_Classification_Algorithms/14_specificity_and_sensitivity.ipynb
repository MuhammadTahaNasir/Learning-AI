## 📝 Core Concepts & Explanations

### Sensitivity and Specificity
- **Sensitivity (TPR):** Ability to correctly spot positive samples.
- **Specificity (TNR):** Ability to correctly spot negative samples ($TN / (TN + FP)$).

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
from sklearn.metrics import confusion_matrix

y_true = np.array([0, 1, 0, 1, 0, 0, 1])
y_pred = np.array([0, 1, 1, 1, 0, 0, 0])

cm = confusion_matrix(y_true, y_pred)
tn, fp, fn, tp = cm.ravel()
print('Sensitivity:', tp / (tp + fn))
print('Specificity:', tn / (tn + fp))
```
