## 📝 Core Concepts & Explanations

### 1. Dictionaries (Hash Maps)
A **dictionary** is an unordered collection of key-value pairs. Keys must be unique and hashable (immutable):
- **Common Operations**:
  - `dict[key]`: Access value (throws error if key missing).
  - `dict.get(key, default)`: Safely access value without raising errors.
  - `dict.items()`: Iterates over `(key, value)` pairs.
  - `dict.keys()`, `dict.values()`: Retrieve collections of keys or values.

### 2. Sets (Unique Elements)
A **set** is an unordered collection of unique elements. Defined using curly braces `{}` or `set()`:
- Sets are mutable, but their elements must be hashable.
- Useful for removing duplicates and calculating mathematical set relationships:
  - Union (`|` or `.union()`)
  - Intersection (`&` or `.intersection()`)
  - Difference (`-` or `.difference()`)
  - Symmetric Difference (`^` or `.symmetric_difference()`)
