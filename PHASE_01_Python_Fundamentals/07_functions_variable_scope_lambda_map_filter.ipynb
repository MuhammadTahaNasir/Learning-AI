## 📝 Core Concepts & Explanations

### 1. Function Scope (LEGB Rule)
Variable lookup resolution in Python follows the **LEGB** order:
- **Local**: Defined inside current function.
- **Enclosing**: Defined in nested/outer functions.
- **Global**: Declared at the top level of the module.
- **Built-in**: Python's pre-loaded keywords (e.g., `len`, `range`).
Use `global x` or `nonlocal x` to modify variables outside local scope.

### 2. Lambda Functions
Anonymous, single-expression functions: `lambda arg1, arg2: expression`.

### 3. Map & Filter
- `map(func, iterable)`: Applies a function to all elements in the sequence.
- `filter(func, iterable)`: Filters items that evaluate to `True` for the function.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
# 1. LEGB Scope demonstration
x = "global"
def outer_func():
    x = "enclosing"
    def inner_func():
        x = "local"
        print("Inner level:", x)
    inner_func()
    print("Outer level:", x)
outer_func()

# 2. Lambda, Map, and Filter
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
evens = list(filter(lambda x: x % 2 == 0, numbers))
print("Squared:", squared)
print("Evens:", evens)
```

1. Write a function `compute_stats(data_list)` that calculates and returns both the sum and the mean of a list of numbers as a tuple.
2. Using `filter()` and `lambda`, extract all words that start with 's' from the list: `words = ['stats', 'probability', 'statistics', 'math', 'sample']`.

```python
# Exercise 1: Sum and Mean
def compute_stats(data_list):
    s = sum(data_list)
    m = s / len(data_list) if data_list else 0
    return s, m
print(compute_stats([10, 20, 30]))

# Exercise 2: Words filtering
words = ['stats', 'probability', 'statistics', 'math', 'sample']
s_words = list(filter(lambda w: w.startswith('s'), words))
print(s_words)
```
