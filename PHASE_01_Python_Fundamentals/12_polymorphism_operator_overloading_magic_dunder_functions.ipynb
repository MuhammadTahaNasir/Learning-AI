## 📝 Core Concepts & Explanations

### 1. Polymorphism
Polymorphism allows different classes to have methods with the same name but different behaviors, allowing a unified interface to process diverse object types.

### 2. Operator Overloading & Magic Methods
Magic/Dunder (Double Underscore) methods allow you to define how custom objects behave with built-in operations (arithmetic, comparisons, length checks):
- `__add__(self, other)`: Overloads addition (`+`)
- `__sub__(self, other)`: Overloads subtraction (`-`)
- `__mul__(self, other)`: Overloads multiplication (`*`)
- `__len__(self)`: Custom return for `len()`
- `__str__(self)`: Human-readable string representation (`print(obj)`)
- `__repr__(self)`: Official string representation for debugging

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
class Vector2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    # Overload addition (+)
    def __add__(self, other):
        return Vector2D(self.x + other.x, self.y + other.y)
        
    # Overload string representations
    def __str__(self):
        return f"Vector({self.x}, {self.y})"
        
    def __len__(self):
        # Custom length: magnitude (integer part)
        return int((self.x**2 + self.y**2)**0.5)

v1 = Vector2D(3, 4)
v2 = Vector2D(1, 2)
v3 = v1 + v2

print("Resulting Vector:", v3)
print("Magnitude of v1:", len(v1))
```
