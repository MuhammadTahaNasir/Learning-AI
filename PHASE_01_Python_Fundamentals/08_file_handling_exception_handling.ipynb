## 📝 Core Concepts & Explanations

### 1. File Handling (Context Managers)
Always use the `with` statement when opening files. It acts as a **context manager**, automatically closing the file stream when execution exits the block, preventing memory leaks:
`with open(file_path, mode) as f:`
Modes: `'r'` (read), `'w'` (write/overwrite), `'a'` (append).

### 2. Exception Handling
Errors are handled dynamically using `try`, `except`, `else`, and `finally` blocks:
- `try`: Block containing code that might throw an error.
- `except ExceptionType`: Code to execute if specific error occurs.
- `else`: Executes if **no** exception occurred in the try block.
- `finally`: Executes unconditionally at the end (useful for closing connections/cleanup).

### 3. Custom Exceptions
Declare custom error structures by inheriting from the base `Exception` class.

2. Write a script that safely attempts to read a non-existent file, catches the `FileNotFoundError`, and prints a user-friendly error message.

```python
# Exercise 1: Custom Exception
class NegativeValueError(Exception):
    pass

def square_root(val):
    if val < 0:
        raise NegativeValueError("Negative numbers are not allowed.")
    return val ** 0.5

try:
    square_root(-4)
except NegativeValueError as e:
    print("Error raised:", e)

# Exercise 2: File Not Found Catch
try:
    with open("ghost.csv", "r") as f:
        data = f.read()
except FileNotFoundError:
    print("Notice: File was not found. Initializing empty dataset.")
```
