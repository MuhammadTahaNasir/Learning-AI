## 📝 Core Concepts & Explanations

### 1. Encapsulation & Access Specifiers
Python doesn't enforce strict access controls like Java/C++, but relies on conventions:
- **Public**: `var`. Accessible from anywhere.
- **Protected**: `_var`. Convention indicating it shouldn't be accessed directly outside subclasses.
- **Private**: `__var`. Restricts external access using Name Mangling.

### 2. Name Mangling
Python renames private attributes prefixing double underscores (e.g. `__attr` in `ClassName` becomes `_ClassName__attr`) to prevent accidental overrides in subclasses.

### 3. Inner (Nested) Classes
Declaring a class inside another class. Useful for logical clustering and nested architectures.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
class Outer:
    def __init__(self):
        self.public_var = "Public"
        self._protected_var = "Protected"
        self.__private_var = "Private"
        self.inner = self.Inner()
        
    class Inner:
        def __init__(self):
            self.data = "Nested Data"
            
obj = Outer()
print("Public variable:", obj.public_var)
print("Protected variable:", obj._protected_var)
try:
    print(obj.__private_var)
except AttributeError as e:
    print("Private access failed:", e)

# Accessing mangled name
print("Mangled access:", obj._Outer__private_var)
print("Inner Class data:", obj.inner.data)
```
