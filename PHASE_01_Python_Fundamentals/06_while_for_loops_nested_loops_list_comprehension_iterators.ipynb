## 📝 Core Concepts & Explanations

### 1. Loop Structures
- `for` loops: Iterate over collections, ranges, or generator expressions.
- `while` loops: Execute as long as a boolean constraint remains `True`.
- Nested loops: Loops within loops (e.g., iterating through a 2D matrix).

### 2. List Comprehension
An elegant, faster syntax to build lists from existing iterables:
`[expression for item in iterable if condition]`

### 3. Iterators & Iterables
- An **iterable** is any object capable of returning its members one at a time (e.g., list, tuple, string).
- An **iterator** is the stream object returned by calling `iter()` on an iterable. It yields items sequentially via `next()`.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
# 1. Nested loops for matrix traversal
matrix = [[1, 2], [3, 4]]
for row in matrix:
    for val in row:
        print(val, end=" ")
print("")

# 2. List Comprehension vs For-Loop
squares = [x**2 for x in range(1, 6)]
print("Squares list comprehension:", squares)

# 3. Iterators
iterable_list = [10, 20]
iterator_obj = iter(iterable_list)
print(next(iterator_obj))
print(next(iterator_obj))
```
