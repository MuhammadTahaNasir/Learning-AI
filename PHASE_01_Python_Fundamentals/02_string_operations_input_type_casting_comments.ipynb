## 📝 Core Concepts & Explanations

### 1. String Slicing & Operations
Strings are sequences of characters. You can access individual characters using **indexing** (0-based) or a range of characters using **slicing**:
`string[start:stop:step]`
- Slicing is *inclusive* of the `start` index and *exclusive* of the `stop` index.
- Negative indexes count backward from the end of the string (e.g., `-1` is the last character).
- Strings are **immutable**; they cannot be altered in-place.

### 2. User Input & Casting
- `input(prompt)` always returns user input as a **string** (`str`).
- To perform math on inputs, you must explicitly cast them using type conversion functions: `int()`, `float()`, or `str()`.

### 3. Comments
- `#` for single-line comments.
- Triple quotes `"""` or `'''` for docstrings and multi-line explanations.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
# 1. Slicing and Concatenation
text = "Machine Learning"
print("First character:", text[0])
print("Last character:", text[-1])
print("Slice [0:7]:", text[0:7])
print("Reverse string:", text[::-1])

# 2. String repetition and checks
print("ML " * 3)
print("Machine" in text)

# 3. Type Casting example
num_str = "123"
num_int = int(num_str)
print("Cast to int:", num_int + 7)
```
