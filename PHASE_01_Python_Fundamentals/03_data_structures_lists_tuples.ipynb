## 📝 Core Concepts & Explanations

### 1. Lists (Mutable Sequences)
A **list** is an ordered, mutable sequence of values. Lists are defined using square brackets `[]`:
- **Mutable**: Elements can be added, changed, or removed in-place.
- **Common Methods**:
  - `.append(x)`: Adds item `x` to the end.
  - `.extend(iterable)`: Appends elements from another iterable.
  - `.insert(i, x)`: Inserts item `x` at index `i`.
  - `.pop(i)`: Removes and returns the item at index `i`.
  - `.remove(x)`: Removes the first occurrence of item `x`.

### 2. Tuples (Immutable Sequences)
A **tuple** is an ordered, immutable sequence of values. Defined using parentheses `()`:
- **Immutable**: Once created, elements cannot be added, changed, or deleted.
- Faster and safer than lists for read-only datasets.
- **Tuple Unpacking**: Assigning individual elements of a tuple to variables in one line.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
# 1. List Mutability and Operations
fruits = ["apple", "banana", "cherry"]
fruits.append("date")
fruits[1] = "blueberry"
print("List after modification:", fruits)

# 2. Tuple Immutability & Unpacking
coordinates = (40.7128, -74.0060)
try:
    coordinates[0] = 34.0522
except TypeError as e:
    print("Error:", e)  # Tuples cannot be modified

# Unpacking coordinates
lat, lon = coordinates
print(f"Latitude: {lat}, Longitude: {lon}")
```

### Exercises

1. Given a list `numbers = [3, 1, 4, 1, 5, 9, 2]`, sort it in descending order and remove duplicates without using sets.
2. Combine two lists `keys = ['name', 'age']` and `values = ['Taha', 21]` into a list of tuples like `[('name', 'Taha'), ('age', 21)]` using `zip()`.

```python
# Exercise 1: Sort and deduplicate list
numbers = [3, 1, 4, 1, 5, 9, 2]
unique_numbers = []
for n in numbers:
    if n not in unique_numbers:
        unique_numbers.append(n)
unique_numbers.sort(reverse=True)
print(unique_numbers)

# Exercise 2: Zip lists
keys = ['name', 'age']
values = ['Taha', 21]
combined = list(zip(keys, values))
print(combined)
```
