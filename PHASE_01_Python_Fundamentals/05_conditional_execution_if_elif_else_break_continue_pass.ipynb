## 📝 Core Concepts & Explanations

### 1. Conditional Branching
Python executes code block conditional branches using `if`, `elif` (else if), and `else` blocks:
- Code blocks are defined via **indentation** (typically 4 spaces).
- Uses comparison operators (`==`, `!=`, `<`, `>`, `<=`, `>=`) and boolean operators (`and`, `or`, `not`).

### 2. Loop Interrupts
- `break`: Terminates the innermost loop immediately.
- `continue`: Skips the rest of the current loop iteration and moves to the next.
- `pass`: A null statement/placeholder used when statement block syntactically requires code but you want no action.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
# 1. Conditional structure example
accuracy = 0.88
if accuracy >= 0.95:
    print("Model performance: Excellent")
elif accuracy >= 0.80:
    print("Model performance: Good")
else:
    print("Model performance: Needs tuning")

# 2. Loop Control Statement
print("Checking loop interrupts:")
for i in range(1, 6):
    if i == 3:
        continue  # Skip 3
    if i == 5:
        break  # Abort loop at 5
    print(i)
```

```python
# Exercise 1: Triangle classifier
a, b, c = 5, 5, 8
if a == b == c:
    print("Equilateral")
elif a == b or b == c or a == c:
    print("Isosceles")
else:
    print("Scalene")

# Exercise 2: Loop with pass
for i in range(1, 11):
    if i % 2 != 0:
        pass
    else:
        print(i)
```
