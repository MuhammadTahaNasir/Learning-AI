## 📝 Core Concepts & Explanations

### 1. Types of Inheritance
- **Single**: Child inherits from a single parent.
- **Multilevel**: Parent -> Child -> Grandchild.
- **Hierarchical**: Multiple subclasses inherit from a single parent.
- **Multiple**: A subclass inherits from more than one parent.

### 2. Method Resolution Order (MRO)
When using multiple inheritance, class namespaces can conflict (the Diamond Problem). Python solves this using the **C3 Linearization** algorithm to determine Method Resolution Order (MRO):
- Call `ClassName.mro()` or print `obj.__class__.__mro__` to view the hierarchy.
- Use `super()` to call parent constructors and methods safely according to MRO sequence.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
class Base:
    def show(self):
        print("Base execution")

class Left(Base):
    def show(self):
        print("Left start")
        super().show()
        print("Left end")

class Right(Base):
    def show(self):
        print("Right start")
        super().show()
        print("Right end")

class SubChild(Left, Right):
    def show(self):
        print("SubChild start")
        super().show()
        print("SubChild end")

child = SubChild()
child.show()
print("\nMRO order:", SubChild.mro())
```
