## 📝 Core Concepts & Explanations

### Built-in Plotting in Pandas
Pandas provides wrapper methods around `matplotlib.pyplot` directly on Series and DataFrames via the `plot` accessor:

- `df['col'].plot(kind='line')`: Line plots (default).
- `df['col'].plot(kind='bar')`: Bar charts for categorical variables.
- `df['col'].plot(kind='hist')`: Histograms for distributions.
- `df.plot(kind='scatter', x, y)`: Scatter plots for bivariate analysis.
- `df['col'].plot(kind='box')`: Box plots for statistical summaries (outliers).

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Generate plot data internally
class_counts = df['Pclass'].value_counts().sort_index()

# Plotting directly with Pandas
class_counts.plot(kind='bar', title='Passenger Counts by Pclass')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
# We close or show depending on notebook context
plt.close() # Prevents showing output during script run
print("Pandas plot setup completed successfully!")
```
