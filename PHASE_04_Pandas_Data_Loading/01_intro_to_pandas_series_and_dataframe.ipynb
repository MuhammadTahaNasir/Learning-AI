## 📝 Core Concepts & Explanations

### 1. What is Pandas?
Pandas is the primary library for data manipulation and analysis in Python. It provides high-performance data structures: **Series** (1-dimensional) and **DataFrame** (2-dimensional).

### 2. Core Data Structures
- **Series**: A 1D labeled array capable of holding any data type (integers, strings, floats, Python objects). It has an axis label called the index.
- **DataFrame**: A 2D labeled data structure with columns of potentially different types, resembling a spreadsheet or SQL table.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
import numpy as np

# 1. Creating a Series
s = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
print("Series:\n", s)

# 2. Creating a DataFrame from a dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Paris', 'London']
}
df = pd.DataFrame(data)
print("\nDataFrame:\n", df)
print("\nColumns:", df.columns)
print("Shape:", df.shape)
```

2. Create a DataFrame from a list of dictionaries representing products (columns: ProductID, Price, Stock).

```python
salaries = pd.Series([50000, 60000, 75000, 80000, 45000], index=['Alex', 'Sarah', 'Ryan', 'John', 'Emma'])
products = pd.DataFrame([
    {'ProductID': 101, 'Price': 15.5, 'Stock': 120},
    {'ProductID': 102, 'Price': 20.0, 'Stock': 80},
    {'ProductID': 103, 'Price': 9.9, 'Stock': 200}
])
print("Salaries Series:\n", salaries)
print("\nProducts DataFrame:\n", products)
```
