## 📝 Core Concepts & Explanations

### Combining Datasets

### 1. Concatenation (`pd.concat`)
Stacks multiple DataFrames vertically (axis=0) or horizontally (axis=1).

### 2. Merging (`pd.merge`)
Database-style joins based on matching keys (columns):
- **Inner Join**: Keeps rows with matching keys in both DataFrames.
- **Outer Join**: Keeps all rows from both DataFrames, filling missing matches with NaN.
- **Left Join**: Keeps all rows from left DataFrame, along with matching rows from the right.
- **Right Join**: Keeps all rows from right DataFrame, along with matching rows from the left.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd

df1 = pd.DataFrame({'id': [1, 2, 3], 'Product': ['Apple', 'Banana', 'Orange']})
df2 = pd.DataFrame({'id': [2, 3, 4], 'Price': [0.5, 0.8, 1.2]})

# 1. Inner Merge
print("Inner Merge:\n", pd.merge(df1, df2, on='id', how='inner'))

# 2. Left Merge
print("\nLeft Merge:\n", pd.merge(df1, df2, on='id', how='left'))

# 3. Concatenation (vertical stacking)
df3 = pd.DataFrame({'id': [5], 'Product': ['Grape']})
print("\nConcatenated DataFrame:\n", pd.concat([df1, df3], axis=0, ignore_index=True))
```
