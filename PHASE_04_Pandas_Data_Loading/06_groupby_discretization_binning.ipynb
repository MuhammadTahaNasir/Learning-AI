## 📝 Core Concepts & Explanations

### Grouping & Discretization (Binning)

### 1. GroupBy
The `groupby` method implements the **Split-Apply-Combine** workflow. It splits data by a category, applies an aggregation (like sum, mean, count), and combines results into a summary table.

### 2. Discretization / Binning
- **`pd.cut`**: Bin values into discrete intervals based on numerical edges (e.g. dividing Age into Child/Adult/Senior).
- **`pd.qcut`**: Bin values into equal-sized quantiles (e.g., dividing salary into quartiles).

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Average Fare by Passenger Class
avg_fare_by_class = df.groupby('Pclass')['Fare'].mean()
print("Average Fare by Pclass:\n", avg_fare_by_class)

# Discretize Age into 3 bins
df_clean = df.dropna(subset=['Age']).copy()
df_clean['AgeGroup'] = pd.cut(df_clean['Age'], bins=[0, 12, 60, 100], labels=['Child', 'Adult', 'Senior'])
print("\nAge group counts:\n", df_clean['AgeGroup'].value_counts())
```
