## 📝 Core Concepts & Explanations

### Data Selection & Indexing in Pandas
We can select subsets of data in a DataFrame using standard indexing or optimized attributes:

- **`df['col_name']`**: Selects a single column as a Series.
- **`df[['col1', 'col2']]`**: Selects multiple columns as a DataFrame.
- **`loc`**: Label-based indexing. Selects rows and columns by their labels: `df.loc[row_label, col_label]`.
- **`iloc`**: Integer position-based indexing. Selects rows and columns by their numerical indices: `df.iloc[row_pos, col_pos]`.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd

# Load sample dataset from URL
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

print("First 3 rows of Iris:\n", df.head(3))

# loc selection (label-based)
print("\nloc selection (row index 0, col 'sepal_length'):", df.loc[0, 'sepal_length'])

# iloc selection (integer position-based)
print("iloc selection (first 2 rows, first 3 columns):\n", df.iloc[:2, :3])
```
