## 📝 Core Concepts & Explanations

### Linear Correlation in Pandas
The correlation coefficient ($r$) measures the strength and direction of the linear relationship between two quantitative variables ($[-1, 1]$):
- $r = 1$: Perfect positive relationship.
- $r = -1$: Perfect negative relationship.
- $r = 0$: No linear relationship.

**`df.corr()`** computes the element-wise correlation matrix for all numerical columns in the DataFrame.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

# Select numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])
print("Correlation Matrix on Iris:\n", numeric_df.corr())
```
