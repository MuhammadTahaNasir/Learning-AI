## 📝 Core Concepts & Explanations

### Handling Missing Data
Real-world data often has missing values (NaN or None). Pandas provides several methods to manage them:

- **Detecting Nulls**: `df.isnull()` returns a boolean mask. `df.isnull().sum()` totals nulls per column.
- **Dropping Nulls (`dropna`)**: Removes rows or columns containing missing values. Useful if missingness is small or column is mostly empty.
- **Imputing Nulls (`fillna`)**: Fills missing values with a placeholder, like a constant value, mean, median, or mode.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Check null status
print("Null counts in Age:", df['Age'].isnull().sum())

# Fill Age missing values with its median
median_age = df['Age'].median()
df_filled = df.copy()
df_filled['Age'] = df_filled['Age'].fillna(median_age)
print("Null counts in Age after fillna:", df_filled['Age'].isnull().sum())

# Drop rows where Cabin is missing
df_dropped = df.dropna(subset=['Cabin'])
print("Shape after dropping Cabin null rows:", df_dropped.shape)
```
