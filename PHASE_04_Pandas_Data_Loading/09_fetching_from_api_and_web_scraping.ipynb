## 📝 Core Concepts & Explanations

### Fetching Remote Web Data
Pandas handles web scraping and API ingestion very efficiently:

### 1. Ingesting APIs
API JSON endpoints can be retrieved using standard `requests` and parsed into a DataFrame: `pd.DataFrame(requests.get(url).json())`.

### 2. Reading HTML Tables
**`pd.read_html(url)`**: Parses all HTML `<table>` elements in the web page into a list of DataFrames. Highly useful for scraping tabular data.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
import requests

# Fetching mock JSON API data
api_url = "https://jsonplaceholder.typicode.com/posts"
response = requests.get(api_url)
if response.status_code == 200:
    posts_df = pd.DataFrame(response.json())
    print("Data ingested from REST API (posts shape):", posts_df.shape)
    print(posts_df.head(2))
```
