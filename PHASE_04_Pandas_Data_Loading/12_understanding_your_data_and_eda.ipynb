## 📝 Core Concepts & Explanations

### Exploratory Data Analysis (EDA)
EDA is the practice of investigating datasets to summarize main patterns, identify anomalies, check assumptions, and visualize distributions. A standard EDA checklist includes:
1. **Dimensions**: Checking shape ($N \times M$).
2. **Datatypes**: Inspecting column data types.
3. **Null values**: Quantifying missing values.
4. **Descriptive Stats**: Summarizing numerical ranges and categorical categories.
5. **Groupings**: Splitting target variable across key categories.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

print("--- 1. Shape of Data ---")
print(df.shape)

print("\n--- 2. Column Types ---")
print(df.dtypes)

print("\n--- 3. Missing Values ---")
print(df.isnull().sum())

print("\n--- 4. Summary Stats ---")
print(df.describe())
```

1. Write an EDA summary report on the Titanic dataset, detailing dimensions, data types, null counts, and mean survival rates grouped by Sex.

```python
titanic_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
t_df = pd.read_csv(titanic_url)
print("Shape:", t_df.shape)
print("Nulls:\n", t_df.isnull().sum())
print("Survival rate by Sex:\n", t_df.groupby('Sex')['Survived'].mean())
```
