## 📝 Core Concepts & Explanations

### Core DataFrame Inspection Methods
To understand a dataset's structure and contents, we use these common operations:

- `df.head(n)` / `df.tail(n)`: View the first or last $n$ rows.
- `df.info()`: Summarizes index, columns, null statuses, and memory usage.
- `df.describe()`: Computes summary statistics (mean, std, percentiles) for numerical columns.
- `col.unique()` / `col.nunique()`: Get unique elements or count of unique elements in a column.
- `col.value_counts()`: Frequency counts of each unique category.
- `df.sort_values(by, ascending)`: Sorts the DataFrame by specific column(s).
- `df.isnull().sum()`: Counts missing values in each column.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("Shape of Titanic:", df.shape)
print("\nUnique values in Survived:", df['Survived'].unique())
print("\nGender value counts:\n", df['Sex'].value_counts())
print("\nMissing values per column:\n", df.isnull().sum())
print("\nFirst 3 rows sorted by Fare descending:\n", df.sort_values(by='Fare', ascending=False).head(3))
```
