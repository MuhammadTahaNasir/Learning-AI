## 📝 Core Concepts & Explanations

### Advanced File Ingestion
When dealing with large datasets, loading options must be tuned to conserve system memory:

- **`usecols`**: Reads only specific columns, avoiding loading unwanted data.
- **`nrows`**: Loads only the first $N$ rows (great for schema inspection).
- **`chunksize`**: Loads data iteratively in blocks of size $N$ as a generator, preventing Out-Of-Memory errors.
- **`parse_dates`**: Automatically parses date strings into datetime objects during loading.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

# 1. Read only Name and Survived
df_small = pd.read_csv(url, usecols=['Name', 'Survived'])
print("Cols read with usecols:", df_small.columns.tolist())

# 2. Chunk loading
chunk_iterator = pd.read_csv(url, chunksize=200)
for i, chunk in enumerate(chunk_iterator, 1):
    print(f"Chunk {i} shape: {chunk.shape}")
```
