## 📝 Core Concepts & Explanations

### Mathematical Derivation
Decomposing expected prediction error into Squared Bias, Variance, and Irreducible Noise:
$$E[(y - \hat{f}(x))^2] = \text{Bias}[\hat{f}(x)]^2 + \text{Var}[\hat{f}(x)] + \sigma^2$$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

np.random.seed(42)
true_f = lambda x: 2 * x + 1.5

x_query = 2.0
predictions = []
for _ in range(50):
    x_train = np.random.uniform(0, 4, 10)
    y_train = true_f(x_train) + np.random.normal(0, 0.5, 10)
    coef, intercept = np.polyfit(x_train, y_train, 1)
    predictions.append(coef * x_query + intercept)

pred_mean = np.mean(predictions)
bias_sq = (pred_mean - true_f(x_query)) ** 2
variance = np.var(predictions)
print(f'Query true value: {true_f(x_query)}')
print(f'Query mean prediction: {pred_mean:.4f}')
print(f'Squared Bias: {bias_sq:.6f}')
print(f'Model Variance: {variance:.6f}')
```
