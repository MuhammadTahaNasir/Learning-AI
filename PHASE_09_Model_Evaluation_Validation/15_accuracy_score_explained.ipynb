## 📝 Core Concepts & Explanations

### Accuracy Math
Accuracy = $\frac{TP + TN}{TP + TN + FP + FN}$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
from sklearn.metrics import accuracy_score
import numpy as np

y_true = np.array([0, 1, 2, 0, 1, 2])
y_pred = np.array([0, 1, 1, 0, 1, 2])
print('Calculated accuracy score:', accuracy_score(y_true, y_pred))
```

### Exercises
1. Express accuracy calculation manually in Python.

```python
manual_acc = np.mean(y_true == y_pred)
print('Manual Accuracy calculation:', manual_acc)
```
