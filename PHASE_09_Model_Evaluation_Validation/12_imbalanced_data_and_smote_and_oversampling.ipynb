## 📝 Core Concepts & Explanations

### Balancing Distributions
- **Oversampling:** Duplicating minority class observations.
- **SMOTE:** Synthetic Minority Over-sampling Technique interpolates new data points along lines joining local neighbors.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=200, n_features=2, n_redundant=0, n_clusters_per_class=1, weights=[0.9, 0.1], random_state=42)

minority_idx = np.where(y == 1)[0]
majority_idx = np.where(y == 0)[0]
num_to_add = len(majority_idx) - len(minority_idx)

np.random.seed(42)
oversampled_noise = np.random.normal(0, 0.1, size=(num_to_add, 2))
synthetic_X = X[np.random.choice(minority_idx, num_to_add)] + oversampled_noise

X_resampled = np.vstack([X, synthetic_X])
y_resampled = np.hstack([y, np.ones(num_to_add)])

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].scatter(X[y==0, 0], X[y==0, 1], color='blue', alpha=0.6, label='Majority')
axes[0].scatter(X[y==1, 0], X[y==1, 1], color='red', alpha=0.6, label='Minority')
axes[0].set_title('Before Oversampling')
axes[0].legend()

axes[1].scatter(X_resampled[y_resampled==0, 0], X_resampled[y_resampled==0, 1], color='blue', alpha=0.6, label='Majority')
axes[1].scatter(X_resampled[y_resampled==1, 0], X_resampled[y_resampled==1, 1], color='red', alpha=0.6, label='Oversampled')
axes[1].set_title('After Oversampling')
axes[1].legend()
plt.show()
```
