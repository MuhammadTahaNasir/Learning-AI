## 📝 Core Concepts & Explanations

### VC Dimension
- **Shattering:** A hypothesis space $H$ shatters a set of points if it can separate all possible label assignments.
- **VC Dimension:** Size of largest set shattered by $H$. Sample complexity bound:
  $$m = O\left(\frac{1}{\epsilon} \log \frac{1}{\delta} + \frac{VC(H)}{\epsilon} \log \frac{1}{\epsilon}\right)$$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

def sample_complexity(vc_dim, epsilon=0.05, delta=0.05):
    return (1.0 / epsilon) * (np.log(1.0 / delta) + vc_dim * np.log(1.0 / epsilon))

print('Required samples for VC=3 (Linear Classifier in 2D):', int(sample_complexity(3)))
```
