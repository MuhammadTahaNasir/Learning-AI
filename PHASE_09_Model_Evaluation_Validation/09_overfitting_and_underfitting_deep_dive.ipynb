### Visualizing Capacity
Varying model parameter complexity (degree of polynomial expansion) changes decision flexibility.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

np.random.seed(42)
X = np.sort(np.random.rand(30) * 5)
y = np.sin(X) + np.random.randn(30) * 0.15
X = X[:, np.newaxis]

degrees = [1, 3, 15]
plt.figure(figsize=(12, 4))
X_plot = np.linspace(0, 5, 100)[:, np.newaxis]

for i, deg in enumerate(degrees):
    model = make_pipeline(PolynomialFeatures(deg), LinearRegression())
    model.fit(X, y)
    plt.subplot(1, 3, i+1)
    plt.scatter(X, y, color='black')
    plt.plot(X_plot, model.predict(X_plot), label=f'Degree {deg}')
    plt.ylim(-1.5, 1.5)
    plt.title(f'Degree {deg} ("Underfit" if deg==1 else "Good" if deg==3 else "Overfit")')
    plt.legend()
plt.tight_layout()
plt.show()
```
