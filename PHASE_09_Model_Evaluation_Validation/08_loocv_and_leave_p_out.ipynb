## 📝 Core Concepts & Explanations

### LOOCV Properties
- **Leave-One-Out (LOOCV):** Fits models using $K=N$. Gives unbiased estimates but high variance.
- **Leave-P-Out:** Leaves $P$ samples for validation.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
from sklearn.model_selection import LeaveOneOut
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
import numpy as np

X, y = load_iris(return_X_y=True)
X = np.vstack([X[:15], X[50:65]])
y = np.hstack([y[:15], y[50:65]])

loo = LeaveOneOut()
scores = []
for train_idx, test_idx in loo.split(X):
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]
    model = LogisticRegression().fit(X_train, y_train)
    scores.append(model.score(X_test, y_test))

print('LOOCV Mean Accuracy:', np.mean(scores))
```
