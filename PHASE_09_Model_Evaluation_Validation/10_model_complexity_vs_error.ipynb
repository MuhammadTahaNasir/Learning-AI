# ✅ Topic 10: Model Complexity vs Error

**Summary:** Train error vs validation error relationship.

---

## 📝 Core Concepts & Explanations

### U-Shaped Curves
As capacity grows, training set error goes to zero, but validation set error begins to rise as generalization fails.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

data = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.3, random_state=42)

depths = range(1, 15)
train_errors = []
test_errors = []

for d in depths:
    clf = DecisionTreeClassifier(max_depth=d, random_state=42).fit(X_train, y_train)
    train_errors.append(1 - clf.score(X_train, y_train))
    test_errors.append(1 - clf.score(X_test, y_test))

plt.figure(figsize=(8, 4))
plt.plot(depths, train_errors, 'b-o', label='Training Error')
plt.plot(depths, test_errors, 'r-o', label='Testing Error')
plt.xlabel('Tree Max Depth (Model Complexity)')
plt.ylabel('Classification Error Rate')
plt.title('Model Complexity vs Train/Test Error')
plt.legend()
plt.grid(True)
plt.show()
```

### Exercises
1. Identify the optimal tree depth range based on the generated plot.

```python
optimal_depth = depths[np.argmin(test_errors)]
print(f'Optimal Depth is at max_depth = {optimal_depth} where validation error is minimized before overfitting begins.')
```
