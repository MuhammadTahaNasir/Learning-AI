## 📝 Core Concepts & Explanations

### Split Visualization
Partitioning dataset into $K$ mutual subsets, using one as validation and $K-1$ as training.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, shuffle=True, random_state=42)
X = np.arange(25)

plt.figure(figsize=(8, 4))
for i, (train_idx, val_idx) in enumerate(kf.split(X)):
    plt.scatter(train_idx, [i]*len(train_idx), color='blue', marker='o', label='Train' if i==0 else "")
    plt.scatter(val_idx, [i]*len(val_idx), color='red', marker='x', label='Validation' if i==0 else "")
plt.yticks(range(5), [f'Fold {i+1}' for i in range(5)])
plt.xlabel('Sample Index')
plt.title('K-Fold Cross-Validation Splitting')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```
