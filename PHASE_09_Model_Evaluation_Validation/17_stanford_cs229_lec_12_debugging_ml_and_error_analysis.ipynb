## 📝 Core Concepts & Explanations

### Learning Diagnostics
- **High Bias (Underfitting):** Train and test errors are high and close. Getting more data will not help.
- **High Variance (Overfitting):** Large gap between train (low) and test (high) errors. Getting more data helps.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve
from sklearn.linear_model import Ridge
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=150, n_features=10, noise=5.0, random_state=42)
train_sizes, train_scores, test_scores = learning_curve(
    Ridge(alpha=10.0), X, y, cv=5, train_sizes=np.linspace(0.1, 1.0, 10), scoring='neg_mean_squared_error', random_state=42
)

train_rmse = np.sqrt(-np.mean(train_scores, axis=1))
test_rmse = np.sqrt(-np.mean(test_scores, axis=1))

plt.figure(figsize=(8, 4))
plt.plot(train_sizes, train_rmse, 'b-o', label='Training Error')
plt.plot(train_sizes, test_rmse, 'r-o', label='Validation Error')
plt.xlabel('Training Set Size')
plt.ylabel('RMSE Error')
plt.title('Learning Curves (Diagnosing Bias vs Variance)')
plt.legend()
plt.grid(True)
plt.show()
```
