## 📝 Core Concepts & Explanations

### Empirical Risk Minimization (ERM)
- **Empirical Risk (Train Error):** $\hat{\epsilon}(h) = \frac{1}{m} \sum_{i=1}^m I(h(x^{(i)}) \neq y^{(i)})$
- **Generalization Error:** $\epsilon(h) = P_{(x,y) \sim D}(h(x) \neq y)$
- **Hoeffding's Inequality:** $P(|\hat{\epsilon}(h) - \epsilon(h)| > \gamma) \le 2e^{-2\gamma^2 m}$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

def hoeffding_bound(m, gamma):
    return 2 * np.exp(-2 * (gamma ** 2) * m)

print('Bound for m=100, gamma=0.1:', hoeffding_bound(100, 0.1))
```
