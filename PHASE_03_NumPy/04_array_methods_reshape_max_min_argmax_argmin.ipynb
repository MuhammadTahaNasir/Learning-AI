## 📝 Core Concepts & Explanations

### Array Manipulation & Statistics

### 1. Reshaping
- `reshape(new_shape)`: Returns an array containing the same data with a new shape. The new shape must have the same total size as the original.

### 2. Statistical Aggregations
- `max()` and `min()`: Find the maximum and minimum values in the array (optionally along a specific axis).
- `argmax()` and `argmin()`: Find the flat indices of the maximum and minimum values.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

arr = np.array([3, 10, 1, 15, 2, 8])
reshaped = arr.reshape((2, 3))
print("Reshaped 2x3 Matrix:\n", reshaped)
print("Max Value:", reshaped.max())
print("Min Value:", reshaped.min())
print("Index of Max:", reshaped.argmax())
print("Index of Max along column (axis=0):", reshaped.argmax(axis=0))
```

### Exercises
1. Create an array of 12 numbers and reshape it into a 3x4 matrix.
2. Find the index of the minimum value in each row (axis=1) of the matrix.

```python
a = np.arange(12).reshape((3, 4))
np.random.shuffle(a) # Shuffle rows to make it interesting
print("Matrix:\n", a)
print("Argmin along rows:", a.argmin(axis=1))
```
