## 📝 Core Concepts & Explanations

### Universal Functions (ufunc)
A universal function (or **ufunc**) is a function that operates on ndarrays in an element-by-element fashion, supporting vectorization.
Common ufuncs include:
- `np.sqrt(arr)`: Computes square root.
- `np.exp(arr)`: Computes natural exponential ($e^x$).
- `np.log(arr)`: Computes natural logarithm ($ln$).
- `np.sin(arr)`, `np.cos(arr)`: Trigonometric operations.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

arr = np.array([1, 4, 9])
print("Square root:", np.sqrt(arr))
print("Exponential:", np.exp(arr))
print("Logarithm (base e):", np.log(arr))
print("Sine (in radians):", np.sin(np.array([0, np.pi/2, np.pi])))
```
