## 📝 Core Concepts & Explanations

### Manipulating Array Elements
Unlike lists, NumPy arrays have a fixed size. Operations like append or delete create a new array copy underneath:

- **Copy (`np.copy`)**: Standard slicing returns a *view* (shallow reference). Modifying a view changes the original array. Use `copy()` to create an independent array.
- **Append (`np.append(arr, values)`)**: Appends values to the end of an array.
- **Insert (`np.insert(arr, obj, values)`)**: Inserts values at a specific index.
- **Delete (`np.delete(arr, obj)`)**: Deletes items at the specified index/indices.
- **Sort (`np.sort(arr)`)**: Returns a sorted copy of the array.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

arr = np.array([40, 10, 30, 20])
view = arr[1:3]
view[0] = 99
print("Original modified by view change:", arr)

copied = np.copy(arr)
copied[0] = 500
print("Original untouched by copy change:", arr)

print("Sorted array:", np.sort(arr))
print("Appended:", np.append(arr, [100, 200]))
print("Deleted index 1:", np.delete(arr, 1))
```

### Exercises
1. Create an array `[10, 5, 8, 3]`. Sort it in descending order, then insert the value `99` at index 2.

```python
a = np.array([10, 5, 8, 3])
sorted_desc = np.sort(a)[::-1]
result = np.insert(sorted_desc, 2, 99)
print("Resulting Array:", result)
```
