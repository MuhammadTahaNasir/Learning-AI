## 📝 Core Concepts & Explanations

### Metadata Attributes of ndarray
Every NumPy array has standard attributes to describe its structure and layout in memory:

- `ndim`: The number of axes (dimensions) of the array.
- `shape`: A tuple of integers showing the size of the array along each dimension.
- `size`: The total number of elements in the array (equal to the product of shape elements).
- `dtype`: An object describing the type of elements in the array (e.g., int32, float64).

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]], dtype=np.float32)
print("Array dimensions (ndim):", arr.ndim)
print("Array shape:", arr.shape)
print("Total elements (size):", arr.size)
print("Data type (dtype):", arr.dtype)
```
