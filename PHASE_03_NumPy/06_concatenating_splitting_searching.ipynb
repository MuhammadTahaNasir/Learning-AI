## 📝 Core Concepts & Explanations

### Structural Merging, Splitting, and Conditional Search

### 1. Concatenation
- `np.concatenate((a1, a2), axis)`: Joins a sequence of arrays along an existing axis.

### 2. Splitting
- `np.split(arr, indices_or_sections)`: Splits an array into multiple sub-arrays.

### 3. Searching
- `np.where(condition)`: Returns indices of elements that satisfy the given condition. Can also be used to conditionally replace values: `np.where(cond, x, y)`.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
print("Concatenate Vertically:\n", np.concatenate((a, b), axis=0))
print("Concatenate Horizontally:\n", np.concatenate((a, b), axis=1))

arr = np.arange(9)
split_arr = np.split(arr, 3)
print("Split array:", split_arr)

# Search even numbers
even_indices = np.where(arr % 2 == 0)
print("Indices of evens:", even_indices)
print("Even values:", arr[even_indices])
```

1. Given an array `arr = np.array([15, 22, 9, 34, 50, 4, 18])`, use `np.where` to replace all values greater than 20 with `999` and all other values with `-1`.

```python
arr = np.array([15, 22, 9, 34, 50, 4, 18])
replaced = np.where(arr > 20, 999, -1)
print("Replaced Array:", replaced)
```
