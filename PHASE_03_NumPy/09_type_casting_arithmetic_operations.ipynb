## 📝 Core Concepts & Explanations

### Arithmetic & Data Type Conversion

### 1. Data Type Casting
- Use `astype(new_dtype)` to cast elements (e.g. converting `float64` to `int32` to save memory).

### 2. Element-Wise Arithmetic
- Basic arithmetic operators `+`, `-`, `*`, `/`, `**` execute element-wise operations on arrays.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

floats = np.array([1.2, 2.5, 3.8])
ints = floats.astype(np.int32)
print("Casted to integers:", ints)

a = np.array([2, 4, 6])
b = np.array([1, 2, 3])
print("Addition:", a + b)
print("Power:", a ** b)
print("Division:", a / b)
```
