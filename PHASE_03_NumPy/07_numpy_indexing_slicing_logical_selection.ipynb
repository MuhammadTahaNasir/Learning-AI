## 📝 Core Concepts & Explanations

### Slicing and Filtering Techniques

### 1. Slice Notation
- `arr[start:stop:step]`
- For 2D matrices: `matrix[row_start:row_stop, col_start:col_stop]`

### 2. Logical (Boolean) Selection
- Creating a boolean condition like `mask = arr > 5` returns an array of booleans.
- Passing this mask back into the array, `arr[mask]`, filters out only `True` values.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

matrix = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

print("Slice row 0 and columns 1-2:", matrix[0, 1:3])
print("Extract 2x2 submatrix:\n", matrix[:2, :2])

# Filter values greater than 50
mask = matrix > 50
print("Boolean Mask:\n", mask)
print("Filtered Values:", matrix[mask])
```

2. Filter out all odd numbers from this matrix using boolean indexing.

```python
m = np.arange(1, 10).reshape((3, 3))
print("Middle column:", m[:, 1])
print("Even numbers:", m[m % 2 == 0])
```
