## 📝 Core Concepts & Explanations

### 1. What is NumPy?
NumPy (Numerical Python) is the fundamental package for scientific computing in Python. It provides a high-performance multidimensional array object, **ndarray**, and tools for working with these arrays.

### 2. Why use NumPy over Lists?
- **Memory Efficiency**: NumPy arrays use contiguous blocks of memory, whereas Python lists store pointers to objects scattered in memory.
- **Performance**: NumPy operations are implemented in compiled C code, allowing element-wise operations (vectorization) without slow Python loops.
- **Convenience**: Math operations on arrays can be written concisely, like `a * b` instead of writing a loop.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import time

# Performance Comparison: List vs NumPy Array
size = 1000000
list1 = list(range(size))
list2 = list(range(size))
arr1 = np.arange(size)
arr2 = np.arange(size)

# Measure list addition time
start = time.time()
list_result = [x + y for x, y in zip(list1, list2)]
print("List time:", time.time() - start, "seconds")

# Measure array addition time
start = time.time()
arr_result = arr1 + arr2
print("NumPy Array time:", time.time() - start, "seconds")
```
