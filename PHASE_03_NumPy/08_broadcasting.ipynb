## 📝 Core Concepts & Explanations

### Broadcasting Rules
Broadcasting allows arithmetic operations on arrays of different shapes. NumPy expands the smaller array to match the size of the larger one without copy overhead.

### Broadcasting Rules Compatibility:
Two dimensions are compatible if:
1. They are equal, OR
2. One of them is 1.

If these conditions are not met, NumPy raises a `ValueError: operands could not be broadcast together`.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

matrix = np.array([[1, 2, 3], [4, 5, 6]]) # shape (2, 3)
row_vector = np.array([10, 20, 30])      # shape (3,)

# Add shape (2,3) to shape (3,) -> Row vector broadcasted across rows
print("Broadcast Addition Result:\n", matrix + row_vector)

col_vector = np.array([[100], [200]])    # shape (2, 1)
print("Col Vector Broadcast:\n", matrix + col_vector)
```
