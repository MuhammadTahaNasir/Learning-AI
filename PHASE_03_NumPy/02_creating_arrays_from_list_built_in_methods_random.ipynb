## 📝 Core Concepts & Explanations

### Array Creation Methods
NumPy provides several convenient built-in functions to initialize arrays:

### 1. From Python Iterables
- `np.array(iterable)`: Converts lists, tuples, etc. into an ndarray.

### 2. Built-in Constant/Range Initializers
- `np.zeros(shape)`: Creates an array filled with 0.
- `np.ones(shape)`: Creates an array filled with 1.
- `np.arange(start, stop, step)`: Creates sequence of numbers with a constant step.
- `np.linspace(start, stop, num)`: Creates a sequence of `num` equally spaced numbers in a closed interval.

### 3. Random Number Generators
- `np.random.rand(d0, d1)`: Uniformly distributed floats in [0, 1).
- `np.random.randn(d0, d1)`: Normal (Gaussian) distributed floats (mean=0, variance=1).
- `np.random.randint(low, high, size)`: Random integers in [low, high).

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

print("Zeros Matrix:\n", np.zeros((2, 3)))
print("Ones Matrix:\n", np.ones((2, 2)))
print("arange(0, 10, 2):", np.arange(0, 10, 2))
print("linspace(0, 1, 5):", np.linspace(0, 1, 5))
print("randint(1, 10, 5):", np.random.randint(1, 10, 5))
```
