## 📝 Core Concepts & Explanations

### Data Classification Schema

```
              DATA TYPES
              /        \
     Structured        Unstructured
     (Tables, CSV)     (Text, Images, Audio)
       /      \ 
 Quantitative  Qualitative (Categorical)
 (Numerical)   |-- Nominal (No order: e.g. Red, Blue)
   |           |-- Ordinal (Ordered: e.g. Low, High)
   |-- Discrete (Counts: e.g. 5 children)
   |-- Continuous (Measurements: e.g. 1.75 meters)
```
- **Structured Data**: Fits in row-column database schemas.
- **Unstructured Data**: Raw media. Needs special preprocessing (feature extraction) to be used in ML.
- **Quantitative**: Numerical data (Discrete vs Continuous).
- **Qualitative**: Categorical data (Nominal vs Ordinal).

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

````python
import pandas as pd
df = pd.DataFrame({
    "Height (Continuous)": [1.75, 1.82, 1.68],
    "Children (Discrete)": [2, 0, 3],
    "Education (Ordinal)": ["BSc", "PhD", "MSc"],
    "City (Nominal)": ["Karachi", "Lahore", "Islamabad"]
})
print(df)
````
