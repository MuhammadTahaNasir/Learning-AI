## 📝 Core Concepts & Explanations

### Bayes' Theorem
Allows us to update our probability beliefs for a hypothesis $H$ given new evidence $E$:
$$P(H | E) = \frac{P(E | H) P(H)}{P(E)}$$

- $P(H | E)$: **Posterior probability** (probability of hypothesis given evidence).
- $P(E | H)$: **Likelihood** (probability of evidence given hypothesis).
- $P(H)$: **Prior probability** (initial probability of hypothesis before evidence).
- $P(E)$: **Marginal likelihood** (normalizing constant): $P(E) = P(E|H)P(H) + P(E|H^c)P(H^c)$.

Crucial for Bayesian Classifiers (Naive Bayes) in classification pipelines.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
# Cancer screening problem
# H: has cancer, E: positive test
# P(H) = 0.01 (prior)
# P(E | H) = 0.95 (sensitivity / likelihood)
# P(E | not H) = 0.05 (false positive rate)

p_h = 0.01
p_e_given_h = 0.95
p_e_given_not_h = 0.05

# P(E) = P(E|H)P(H) + P(E|not H)P(not H)
p_e = (p_e_given_h * p_h) + (p_e_given_not_h * (1 - p_h))

p_h_given_e = (p_e_given_h * p_h) / p_e
print("Posterior Probability P(Cancer | Positive Test):", round(p_h_given_e, 4))
```

### Exercises
1. A spam filter detects 99% of spam emails. 1% of normal emails are falsely marked as spam. If 10% of all incoming emails are spam, what is the probability that an email marked as spam is actually spam?

```python
p_spam = 0.10
p_s_given_spam = 0.99
p_s_given_ham = 0.01
p_s = (p_s_given_spam * p_spam) + (p_s_given_ham * (1 - p_spam))
p_spam_given_s = (p_s_given_spam * p_spam) / p_s
print("P(Spam | Marked Spam):", round(p_spam_given_s, 4))  # ~91.6%
```
