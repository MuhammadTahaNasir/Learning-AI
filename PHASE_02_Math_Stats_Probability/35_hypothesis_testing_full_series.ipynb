## 📝 Core Concepts & Explanations

### Hypothesis Testing Framework

### 1. Formulate Hypotheses
- **Null Hypothesis ($H_0$)**: Statement of no effect or no difference.
- **Alternative Hypothesis ($H_a$)**: What we want to prove.

### 2. Error Types
- **Type I Error ($\alpha$)**: Rejecting $H_0$ when it is actually true (false positive).
- **Type II Error ($\beta$)**: Failing to reject $H_0$ when it is actually false (false negative).

### 3. Z-test vs T-test
- **Z-test**: Used when sample size is large ($n \ge 30$) or population variance $\sigma$ is known.
- **T-test**: Used when sample size is small ($n < 30$) and population variance is unknown (uses sample standard deviation $s$).

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import scipy.stats as stats

# One-sample t-test
# H0: mean = 100
# Ha: mean != 100
scores = [102, 105, 98, 103, 101, 105, 99]
t_stat, p_value = stats.ttest_1samp(scores, 100)
print("T-statistic:", t_stat)
print("p-value:", p_value)
```

### Exercises
1. If significance level $\alpha = 0.05$, do we reject the null hypothesis $H_0$ based on the p-value computed above?

```python
alpha = 0.05
print("Reject Null?", p_value < alpha)  # If True, reject
```
