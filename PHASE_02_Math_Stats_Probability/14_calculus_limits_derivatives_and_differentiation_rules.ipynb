## 📝 Core Concepts & Explanations

### 1. Limits and Derivatives
The derivative of $f(x)$ measures the instantaneous rate of change of $f$ with respect to $x$:
$$f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h}$$

### 2. Standard Rules of Differentiation
- **Power Rule**: $\frac{d}{dx}[x^n] = n x^{n-1}$
- **Constant Factor**: $\frac{d}{dx}[c \cdot f(x)] = c \cdot f'(x)$
- **Sum/Difference Rule**: $\frac{d}{dx}[f(x) \pm g(x)] = f'(x) \pm g'(x)$
- **Product Rule**: $\frac{d}{dx}[f(x)g(x)] = f'(x)g(x) + f(x)g'(x)$
- **Quotient Rule**: $\frac{d}{dx}[\frac{f(x)}{g(x)}] = \frac{f'(x)g(x) - f(x)g'(x)}{(g(x))^2}$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
# Numerical differentiation in python
def f(x):
    return x**2 + 3*x

def derivative(func, x, h=1e-5):
    return (func(x + h) - func(x)) / h

print("Numerical derivative at x=2:", derivative(f, 2))
print("Analytical derivative at x=2 (2x + 3):", 2*2 + 3)
```
