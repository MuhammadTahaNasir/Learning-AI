## 📝 Core Concepts & Explanations

### Estimation Theory

### 1. Point Estimate
A single value calculated from sample data that serves as the best guess for an unknown population parameter.
- **Example**: The sample mean $\bar{x}$ is a point estimate of the population mean $\mu$.

### 2. Interval Estimate (Confidence Interval)
A range of values, derived from sample statistics, that is likely to contain the population parameter.
- Form: $\text{Point Estimate} \pm \text{Margin of Error}$.
