## 📝 Core Concepts & Explanations

### Frequentist vs Bayesian Views
- **Frequentist**: Probability represents the limit of relative frequency in infinite repeated trials. Parameters are fixed but unknown.
- **Bayesian**: Probability measures degrees of belief or certainty. Parameters are treated as random variables with distributions.

### Conjugate Priors
In Bayesian inference, if the posterior distribution is in the same probability family as the prior distribution, the prior is called a **conjugate prior** for the likelihood.
- **Example**: Beta prior is conjugate to Binomial likelihood (Beta-Binomial update: $\alpha_{post} = \alpha_{prior} + \text{successes}$, $\beta_{post} = \beta_{prior} + \text{failures}$).

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
# Beta-Binomial Conjugate Update
# Prior: Beta(2, 2) (Uniform-ish)
# Experiment: 8 successes, 2 failures
prior_alpha, prior_beta = 2, 2
successes, failures = 8, 2

post_alpha = prior_alpha + successes
post_beta = prior_beta + failures
print(f"Posterior Distribution parameters: Beta({post_alpha}, {post_beta})")
```
