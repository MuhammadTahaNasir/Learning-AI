# 📊 Topic 30: Univariate, Bivariate, Multivariate Analysis

**Summary:** Exploring datasets by analyzing one, two, or multiple variables simultaneously.

---

## 📝 Core Concepts & Explanations

### Exploratory Data Analysis (EDA) Dimensions

### 1. Univariate Analysis
Analyzing a single variable in isolation. Looks at shape, central tendency, frequency.
- **Plots**: Histograms, box plots, count plots.

### 2. Bivariate Analysis
Analyzing the relationship between two variables.
- **Plots**: Scatter plots, line plots, bar plots, cross-tabs.

### 3. Multivariate Analysis
Analyzing relationships between three or more variables simultaneously.
- **Plots**: Pair plots, 3D scatter plots, heatmaps of correlation matrices.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate dummy dataset
np.random.seed(42)
df = pd.DataFrame({
    "Size": np.random.normal(150, 20, 100),
    "Price": np.random.normal(300, 50, 100),
    "Rooms": np.random.choice([2, 3, 4], 100)
})

# Bivariate Scatter Plot
sns.scatterplot(data=df, x="Size", y="Price", hue="Rooms", palette="viridis")
plt.title("Bivariate/Multivariate Plot")
plt.show()
```

### Exercises
1. Use `sns.heatmap` to plot the correlation matrix of the generated dataframe `df`.

```python
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.show()
```
