## 📝 Core Concepts & Explanations

### Chi Square Test ($\chi^2$)
Used to analyze categorical variables:

### 1. Goodness-of-Fit Test
Determines if an observed frequency distribution matches an expected distribution.

### 2. Test of Independence
Determines whether two categorical variables are independent or associated:
$$\chi^2 = \sum \frac{(O - E)^2}{E}$$
where $O$ is observed frequencies, $E$ is expected frequencies.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import scipy.stats as stats
import numpy as np

# Contingency table: Columns=Smoking (Yes/No), Rows=Exercise (Low/High)
observed = np.array([[10, 30], [20, 40]])
chi2, p_val, dof, expected = stats.chi2_contingency(observed)
print("Chi2 Statistic:", chi2)
print("p-value:", p_val)
```
