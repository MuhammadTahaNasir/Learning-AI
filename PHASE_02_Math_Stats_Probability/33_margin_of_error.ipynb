## 📝 Core Concepts & Explanations

### Margin of Error (MoE)
Represents the maximum expected difference between the sample statistic and the true population parameter:
$$MoE = z^* \times \left(\frac{\sigma}{\sqrt{n}}\right)$$
where:
- $z^*$ is the critical value (z-score matching confidence level).
- $\frac{\sigma}{\sqrt{n}}$ is the standard error.
- $n$ is the sample size.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
# Calculate MoE for 95% confidence (z*=1.96), sample size n=100, std=5
import math
z_critical = 1.96
std_dev = 5
n = 100
moe = z_critical * (std_dev / math.sqrt(n))
print("Margin of Error:", moe)
```
