## 📝 Core Concepts & Explanations

### 1. Critical Points (Local Maxima/Minima)
A critical point of $f(x)$ occurs where $f'(x) = 0$:
- **Local Minimum**: $f'(x) = 0$ and $f''(x) > 0$ (concave up).
- **Local Maximum**: $f'(x) = 0$ and $f''(x) < 0$ (concave down).

### 2. Convexity
A function $f(x)$ is **convex** if the line segment between any two points on its graph lies above or on the graph. 
- In multi-variable calculus, $f$ is convex if its Hessian matrix $H$ (second derivative matrix) is positive semi-definite.
- **Convex Optimization**: Crucial in ML because any local minimum is guaranteed to be the global minimum (e.g. Linear Regression, SVM).

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
# Plotting a convex function f(x) = x^2
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = x**2

plt.plot(x, y, label='f(x) = x^2 (Convex)')
plt.scatter([0], [0], color='red', zorder=5, label='Global Min (0, 0)')
plt.title("Convex Optimization Space")
plt.legend()
plt.show()
```
