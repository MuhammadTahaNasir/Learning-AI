# 📊 Topic 20: Variance and Standard Deviation

**Summary:** Measures of data dispersion: Variance and Standard Deviation formulations for populations and samples.

---

## 📝 Core Concepts & Explanations

### 1. Population Variance ($\sigma^2$)
Average of squared differences from the Mean:
$$\sigma^2 = \frac{1}{N} \sum_{i=1}^N (x_i - \mu)^2$$

### 2. Sample Variance ($s^2$)
Uses Bessel's correction ($N-1$ in denominator) to provide an unbiased estimate:
$$s^2 = \frac{1}{N-1} \sum_{i=1}^n (x_i - \bar{x})^2$$

### 3. Standard Deviation ($\sigma$ or $s$)
The square root of the variance, returning dispersion to the original measurement units:
$$\sigma = \sqrt{\sigma^2}$$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
data = [2, 4, 4, 6, 8]

print("Population Variance:", np.var(data))
print("Sample Variance (ddof=1):", np.var(data, ddof=1))
print("Standard Deviation:", np.std(data))
```
