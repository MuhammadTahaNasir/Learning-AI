**Summary:** Core paradigms of machine learning: learning from labeled data, finding patterns, or taking action in environments.

---

## 📝 Core Concepts & Explanations

### 1. Supervised Learning
Learning with a teacher. The dataset consists of input-output pairs $(X, y)$. Goal: learn a mapping function $f: X \rightarrow y$.
- **Regression**: Continuous output (e.g. predicting price).
- **Classification**: Discrete labels (e.g. predicting spam/ham).

### 2. Unsupervised Learning
Learning without labels. Dataset consists only of inputs $X$. Goal: discover underlying structures or distributions.
- **Clustering**: Grouping similar items (e.g. customer segmentation).
- **Dimensionality Reduction**: Projecting data to fewer dimensions (e.g. PCA).

### 3. Reinforcement Learning (RL)
Learning by trial and error. An **agent** interacts with an **environment** by performing **actions** to maximize cumulative **rewards**.
