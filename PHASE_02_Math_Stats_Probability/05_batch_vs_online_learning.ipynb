# 📊 Topic 5: Batch vs Online Learning

**Summary:** Model updating strategies: offline batch training vs incremental updates from streaming data.

---

## 📝 Core Concepts & Explanations

### 1. Batch (Offline) Learning
The model is trained on the entire dataset at once. 
- If new data arrives, you must retrain a completely new model from scratch using old and new data.
- Computationally heavy; updates happen infrequently.

### 2. Online (Incremental) Learning
The model is trained incrementally by feeding it data instances sequentially, either individually or in small groups (mini-batches).
- Rapid adaptation to new data streams.
- Low memory footprint.
- **Out-of-Core Learning**: Training on datasets that do not fit into RAM.
