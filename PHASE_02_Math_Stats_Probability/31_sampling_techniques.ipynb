## 📝 Core Concepts & Explanations

### Probability Sampling Techniques
1. **Simple Random Sampling (SRS)**: Every member of the population has an equal chance of selection.
2. **Stratified Sampling**: Population is divided into subgroups (strata), and SRS is applied to each stratum (critical for ML train/test splits to avoid class imbalance biases).
3. **Systematic Sampling**: Members are selected at regular intervals (e.g. every $k$-th item).
4. **Cluster Sampling**: Population is divided into clusters, and random clusters are chosen entirely.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
df = pd.DataFrame({
    "ID": range(1, 11),
    "Class": ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B', 'B']
})

# Stratified Sample using pandas groupby
stratified_sample = df.groupby('Class', group_keys=False).apply(lambda x: x.sample(2))
print("Stratified Sample (2 from each class):\n", stratified_sample)
```
