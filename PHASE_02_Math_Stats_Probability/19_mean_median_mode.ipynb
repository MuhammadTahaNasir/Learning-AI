## 📝 Core Concepts & Explanations

### Measures of Central Tendency

### 1. Mean (Arithmetic Average)
$$\mu = \frac{1}{N} \sum_{i=1}^N x_i$$
- Sensitive to extreme outliers.

### 2. Median (Middle Value)
The middle value when data is sorted in ascending order. If $N$ is even, it is the average of the two middle values.
- **Robust to outliers**.

### 3. Mode (Most Frequent)
The value that occurs most frequently in the dataset.
- Useful for categorical data.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
from scipy import stats

dataset = [1, 2, 2, 3, 4, 100]  # 100 is an outlier

print("Mean:", np.mean(dataset))
print("Median:", np.median(dataset))  # Much more representative middle
print("Mode:", stats.mode(dataset, keepdims=True).mode[0])
```
