## 📝 Core Concepts & Explanations

### 1. Determinants
The determinant $\det(A)$ of a square matrix describes its scaling factor. If $\det(A) = 0$, the matrix is singular (no inverse).

### 2. Orthogonality
Two vectors are orthogonal if their dot product is zero: $\mathbf{u} \cdot \mathbf{v} = 0$.

### 3. Eigenvalues & Eigenvectors
For a square matrix $A$, a non-zero vector $\mathbf{v}$ is an **eigenvector** if:
$$A\mathbf{v} = \lambda\mathbf{v}$$
where $\lambda$ is a scalar called the **eigenvalue**.
- Critical for Dimensionality Reduction (PCA) and Spectral Clustering.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

A = np.array([[4, 2], [1, 3]])
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
print("Determinant of A:", np.linalg.det(A))
```

### Exercises
1. Define a symmetric matrix $S = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}$ and compute its eigenvalues. Verify that eigenvectors of symmetric matrices are orthogonal.

```python
S = np.array([[2, 1], [1, 2]])
vals, vecs = np.linalg.eig(S)
# vecs[:, 0] is first eigenvector, vecs[:, 1] is second
dot_prod = np.dot(vecs[:, 0], vecs[:, 1])
print("Dot product of eigenvectors (orthogonal):", np.round(dot_prod, 5))
```
