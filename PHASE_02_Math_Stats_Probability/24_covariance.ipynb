## 📝 Core Concepts & Explanations

### Covariance
Covariance measures the joint variability of two random variables:
$$\text{Cov}(X, Y) = \frac{1}{N} \sum_{i=1}^N (x_i - \mu_x)(y_i - \mu_y)$$
- **Positive Covariance**: As $X$ increases, $Y$ tends to increase.
- **Negative Covariance**: As $X$ increases, $Y$ tends to decrease.
- Unlike correlation, covariance is **scale-dependent** (changing data units changes the covariance value).

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

cov_matrix = np.cov(x, y, ddof=0)
print("Covariance Matrix:\n", cov_matrix)
print("Cov(x, y):", cov_matrix[0, 1])
```
