## 📝 Core Concepts & Explanations

### Common Challenges
1. **Insufficient Quantity of Data**: Models need thousands of examples to learn patterns.
2. **Nonrepresentative Training Data**: Leads to sampling bias.
3. **Poor Quality Data**: Noisy data, outliers, or missing features degrade model capacity.
4. **Irrelevant Features**: 'Garbage in, garbage out'. Requires Feature Selection.
5. **Overfitting**: Model performs well on training data but poorly on test data. Fixed by regularization, simple models, or more data.
6. **Underfitting**: Model is too simple to learn the underlying structure. Fixed by complex models or better features.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import matplotlib.pyplot as plt
import numpy as np

# Visualizing underfitting vs overfitting conceptually
x = np.linspace(-3, 3, 20)
y = 0.5 * x**3 - x + np.random.normal(0, 0.5, 20)

plt.scatter(x, y, color='red', label='Data points')
plt.plot(x, 0.5*x, label='Underfit (Linear)', color='blue')
plt.plot(x, 0.5*x**3 - x, label='Perfect Fit (Cubic)', color='green')
plt.legend()
plt.title("Concept: Underfitting vs Perfect Fit")
plt.show()
```
