---

## 📝 Core Concepts & Explanations

### 1. Coefficient of Variation (CV)
Relative standard deviation (dimensionless): $CV = \frac{\sigma}{\mu}$. Used to compare variance across different scales.

### 2. Z Score
Measures how many standard deviations a data point $x$ is from the mean:
$$z = \frac{x - \mu}{\sigma}$$
- Used to detect outliers (typically $|z| > 3$).

### 3. Percentiles and Quartiles
- **Percentile**: The value below which a percentage of data falls (e.g. 90th percentile).
- **Quartiles**: Splits sorted data into four equal groups:
  - $Q_1$: 25th percentile (lower quartile).
  - $Q_2$: 50th percentile (Median).
  - $Q_3$: 75th percentile (upper quartile).

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
data = [10, 12, 14, 15, 18, 20, 22, 100]  # 100 is outlier

# Calculate Z scores
mean = np.mean(data)
std = np.std(data)
z_scores = [(x - mean) / std for x in data]
print("Z scores:", np.round(z_scores, 2))

# Percentiles
print("25th percentile (Q1):", np.percentile(data, 25))
print("75th percentile (Q3):", np.percentile(data, 75))
```
