## 📝 Core Concepts & Explanations

### 1. Matrix Transpose
The transpose $A^T$ of matrix $A$ swaps rows with columns: $(A^T)_{ij} = A_{ji}$.
Properties: $(AB)^T = B^T A^T$.

### 2. Systems of Linear Equations
A system of equations can be written as:
$$A\mathbf{x} = \mathbf{b}$$

### 3. Matrix Inverse
If $A$ is square, its inverse $A^{-1}$ satisfies:
$$A A^{-1} = A^{-1} A = I$$
where $I$ is the Identity matrix. If $\det(A) \neq 0$, the inverse exists, and:
$$\mathbf{x} = A^{-1}\mathbf{b}$$
In ML, inverses are crucial for solving Normal Equations in Linear Regression.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

A = np.array([[2, 1], [1, 3]])
b = np.array([5, 5])

# Solve Ax = b
x = np.linalg.solve(A, b)
print("Solution x:", x)

# Transpose and Inverse
print("A Transpose:\n", A.T)
print("A Inverse:\n", np.linalg.inv(A))
```

### Exercises
1. Solve the system $2x + y = 8$, $x - y = 1$ using matrix inverse.

```python
A = np.array([[2, 1], [1, -1]])
b = np.array([8, 1])
x = np.dot(np.linalg.inv(A), b)
print("x, y:", x)  # x=3, y=2
```
