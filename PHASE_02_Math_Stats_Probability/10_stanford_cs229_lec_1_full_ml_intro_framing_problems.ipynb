## 📝 Core Concepts & Explanations

### CS229 Lecture 1 Takeaways
- **Supervised Learning**: Learn mapping function from $X$ to $y$. Example: housing prediction.
- **Unsupervised Learning**: Learn structures, like grouping. Example: microarray gene grouping, social network analysis.
- **Reinforcement Learning**: Uses rewards. Example: helicopter autopilot.
- **Problem Formulation**: Choose features $x^{(i)}$ and target labels $y^{(i)}$ carefully. Ensure the data is clean and matches the task objective.
