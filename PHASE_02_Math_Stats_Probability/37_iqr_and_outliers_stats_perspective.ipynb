## 📝 Core Concepts & Explanations

### Outlier Detection using IQR
The **Interquartile Range (IQR)** measures statistical dispersion:
$$IQR = Q_3 - Q_1$$

### Outlier Boundaries
Data points are classified as outliers if they fall outside the boundaries:
- **Lower Bound**: $Q_1 - 1.5 \times IQR$
- **Upper Bound**: $Q_3 + 1.5 \times IQR$
- In Box Plots, the whiskers represent these limits.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
data = [3, 5, 6, 7, 7, 8, 9, 25]  # 25 is outlier

q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = [x for x in data if x < lower_bound or x > upper_bound]
print(f"IQR: {iqr}, Outlier Limits: [{lower_bound}, {upper_bound}]")
print("Outliers identified:", outliers)
```

### Exercises
1. Write a function `remove_outliers_iqr(dataset)` that takes a list, computes the IQR boundaries, and returns a new list with outliers removed.

```python
def remove_outliers_iqr(dataset):
    arr = np.array(dataset)
    q1, q3 = np.percentile(arr, [25, 75])
    iqr = q3 - q1
    filtered = arr[(arr >= q1 - 1.5*iqr) & (arr <= q3 + 1.5*iqr)]
    return list(filtered)

print("Deduplicated dataset:", remove_outliers_iqr([3, 5, 6, 7, 7, 8, 9, 25]))
```
