## 📝 Core Concepts & Explanations

### 1. Partial Derivatives
For a function of multiple variables $f(x, y)$, the partial derivative $\frac{\partial f}{\partial x}$ is computed by treating all other variables (like $y$) as constants.

### 2. The Gradient
The **gradient** $\nabla f(\mathbf{x})$ is a vector of all partial derivatives:
$$\nabla f(x, y) = \begin{bmatrix} \frac{\partial f}{\partial x} \\ \frac{\partial f}{\partial y} \end{bmatrix}$$
- The gradient points in the direction of steepest ascent.
- Used to update weights in Gradient Descent: $\mathbf{w} \leftarrow \mathbf{w} - \alpha \nabla L(\mathbf{w})$.

### 3. Chain Rule
For nested functions $y = f(g(x))$:
$$\frac{dy}{dx} = f'(g(x)) \cdot g'(x)$$
Critical for backpropagation in neural networks.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
# Compute numerical gradient of f(x, y) = x^2 + 3xy + y^2 at (1, 2)
def f(x, y):
    return x**2 + 3*x*y + y**2

h = 1e-5
grad_x = (f(1 + h, 2) - f(1, 2)) / h
grad_y = (f(1, 2 + h) - f(1, 2)) / h

print("Numerical Gradient at (1, 2):", [grad_x, grad_y])
print("Analytical Gradient at (1, 2) [2x+3y, 3x+2y]:", [2*1 + 3*2, 3*1 + 2*2])
```
