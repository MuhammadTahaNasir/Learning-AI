**Summary:** How ML systems generalize from training data to predict unseen data.

---

## 📝 Core Concepts & Explanations

### 1. Instance-Based Learning
The system learns the training examples by heart, then generalizes to new cases by comparing them using a similarity metric (e.g. Euclidean distance).
- **Example**: k-Nearest Neighbors (k-NN).
- No training time (lazy learner), but high inference time.

### 2. Model-Based Learning
The system builds a model (parameterized mathematical equation) from the training data, then uses that model to make predictions.
- **Example**: Linear Regression ($y = w_1 x + w_0$).
- Long training time (eager learner) to optimize weights, but instant inference time.
