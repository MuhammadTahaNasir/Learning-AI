## 📝 Core Concepts & Explanations

### 1. Venn Diagram Hierarchy
```
+--------------------------------------------------+
| Artificial Intelligence (AI)                     |
|   (Programs that mimic human intelligence)        |
|                                                  |
|   +------------------------------------------+   |
|   | Machine Learning (ML)                    |   |
|   |   (Algorithms that learn from data)      |   |
|   |                                          |   |
|   |   +----------------------------------+   |   |
|   |   | Deep Learning (DL)               |   |   |
|   |   |   (Multi-layer Neural Networks)  |   |   |
|   |   +----------------------------------+   |   |
|   +------------------------------------------+   |
+--------------------------------------------------+
```

### 2. Definitions
- **AI**: Broadest term. Covers anything simulating human intelligence (e.g. expert systems, pathfinding).
- **ML**: Subset of AI. Systems that learn parameters from data instead of hardcoded rules.
- **DL**: Subset of ML. Uses Deep Neural Networks with many hidden layers to automatically extract features from raw data (e.g. pixels, audio).
