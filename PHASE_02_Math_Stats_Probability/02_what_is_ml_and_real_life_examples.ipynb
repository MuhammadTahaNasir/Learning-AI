### Real Life Examples of ML
ML is embedded in many modern services:
- **Spam Filtering**: Classifying emails as Spam or Ham based on words.
- **Recommendation Systems**: Netflix/YouTube recommending videos using your watch history.
- **Predictive Maintenance**: Predicting when a factory machine will fail based on sensor data.
- **Financial Fraud Detection**: Flagging suspicious transactions based on spending behavior patterns.
