## 📝 Core Concepts & Explanations

### Probability Foundations
- **Sample Space ($S$)**: Set of all possible outcomes (e.g. for a coin toss: $\{H, T\}$).
- **Event ($A$)**: A subset of the sample space.
- **Probability of Event ($P(A)$)**:
  $$0 \le P(A) \le 1$$
  $$P(S) = 1$$
- **Complement Rule**: $P(A^c) = 1 - P(A)$.
- **Addition Rule**: $P(A \cup B) = P(A) + P(B) - P(A \cap B)$.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
# Simulate roll of a fair die
import numpy as np
rolls = np.random.randint(1, 7, size=10000)
prob_even = np.sum(rolls % 2 == 0) / len(rolls)
print("Simulated Probability of rolling an even number:", prob_even)
```
