# 📊 Topic 22: Skewness and Kurtosis

**Summary:** Measures of distribution shape: asymmetry (skewness) and tail weight (kurtosis).

---

## 📝 Core Concepts & Explanations

### 1. Skewness (Measure of Asymmetry)
- **Positive (Right) Skew**: Tail extends to the right. $\text{Mean} > \text{Median} > \text{Mode}$.
- **Negative (Left) Skew**: Tail extends to the left. $\text{Mean} < \text{Median} < \text{Mode}$.
- **Symmetric**: Skewness $\approx 0$.

### 2. Kurtosis (Measure of Tailedness)
Measures the thickness/weight of the tails of a distribution relative to a Normal distribution:
- **Mesokurtic**: Kurtosis $= 3$ (Normal distribution).
- **Leptokurtic**: Kurtosis $> 3$ (heavy tails, high peak).
- **Platykurtic**: Kurtosis $< 3$ (thin tails, flat peak).

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
from scipy.stats import skew, kurtosis
import numpy as np

# Generate asymmetric right-skewed data
data = np.random.exponential(scale=2, size=1000)
print("Skewness:", skew(data))
print("Kurtosis:", kurtosis(data))  # excess kurtosis (kurtosis - 3)
```
