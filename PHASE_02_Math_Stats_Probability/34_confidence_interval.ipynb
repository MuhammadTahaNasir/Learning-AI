## 📝 Core Concepts & Explanations

### Confidence Interval (CI)
A confidence interval specifies the range of values within which a population parameter is expected to fall with a certain probability (confidence level, e.g. 95%):
$$CI = \bar{x} \pm z^* \left(\frac{s}{\sqrt{n}}\right)$$
- **95% Confidence interpretation**: If we take 100 independent samples and construct a 95% CI for each, approximately 95 of those intervals will contain the true population mean.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import scipy.stats as stats

data = [12, 15, 14, 16, 18, 15, 13, 14, 15, 16]
mean = np.mean(data)
sem = stats.sem(data)  # standard error of the mean
ci = stats.t.interval(0.95, df=len(data)-1, loc=mean, scale=sem)
print("95% Confidence Interval for mean:", ci)
```

### Exercises

```python
ci_99 = stats.t.interval(0.99, df=len(data)-1, loc=mean, scale=sem)
print("99% Confidence Interval:", ci_99)
```
