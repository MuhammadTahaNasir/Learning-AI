## 📝 Core Concepts & Explanations

### 1. Vectors & Dot Product
A vector $\mathbf{x} \in \mathbb{R}^n$ is an ordered sequence of numbers. 
The **dot product** of two vectors $\mathbf{u}, \mathbf{v} \in \mathbb{R}^n$ is:
$$\mathbf{u} \cdot \mathbf{v} = \sum_{i=1}^n u_i v_i = \mathbf{u}^T \mathbf{v}$$

### 2. Vector Norms
Norms measure vector magnitude (size):
- **$L_1$ Norm (Manhattan)**: $\|\mathbf{x}\|_1 = \sum |x_i|$
- **$L_2$ Norm (Euclidean)**: $\|\mathbf{x}\|_2 = \sqrt{\sum x_i^2}$

### 3. Matrix Multiplication
If $A$ is $m \times n$ and $B$ is $n \times p$, then $C = AB$ is $m \times p$:
$$C_{ij} = \sum_{k=1}^n A_{ik} B_{kj}$$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

u = np.array([1, 2, 3])
v = np.array([4, 5, 6])

print("Dot Product:", np.dot(u, v))
print("L1 Norm of u:", np.linalg.norm(u, 1))
print("L2 Norm of u:", np.linalg.norm(u, 2))

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print("Matrix Multiplication AB:\n", np.dot(A, B))
```
