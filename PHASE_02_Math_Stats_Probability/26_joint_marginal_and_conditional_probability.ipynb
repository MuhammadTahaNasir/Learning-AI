## 📝 Core Concepts & Explanations

### Probabilities in Multi-Event Spaces

### 1. Joint Probability
The probability of two events occurring simultaneously: $P(A \cap B)$ or $P(A, B)$.

### 2. Marginal Probability
The probability of a single event occurring, ignoring other variables:
$$P(A) = \sum_b P(A, B=b)$$

### 3. Conditional Probability
The probability of event $A$ occurring given that event $B$ has already occurred:
$$P(A | B) = \frac{P(A \cap B)}{P(B)}$$
- **Independent Events**: If $A$ and $B$ are independent, $P(A | B) = P(A)$, and $P(A \cap B) = P(A)P(B)$.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
# Simple contingency table simulation
# rows: Gender (0=Male, 1=Female), cols: Prefers Python (0=No, 1=Yes)
table = np.array([[30, 20], [10, 40]])  # sum = 100

joint_prob_female_python = table[1, 1] / 100  # P(F, Yes)
marginal_prob_python = np.sum(table[:, 1]) / 100  # P(Yes)
conditional_female_given_python = table[1, 1] / np.sum(table[:, 1])  # P(F | Yes)

print("Joint Probability P(Female, Yes):", joint_prob_female_python)
print("Marginal Probability P(Yes):", marginal_prob_python)
print("Conditional Probability P(Female | Yes):", conditional_female_given_python)
```
