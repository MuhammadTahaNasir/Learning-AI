## 📝 Core Concepts & Explanations

### Locally Weighted Regression (LWR)
A non-parametric algorithm: weights $w^{(i)}$ are evaluated centered around query prediction point $x$:
$$w^{(i)} = \exp\left(-\frac{(x^{(i)} - x)^2}{2\tau^2}\right)$$
Cost to minimize: $\sum w^{(i)} (y^{(i)} - \theta^T x^{(i)})^2$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

# Generate noisy 1D sine wave
np.random.seed(0)
X = np.sort(np.random.rand(50) * 6)
y = np.sin(X) + np.random.randn(50) * 0.1

def lwr_predict(x_query, X, y, tau):
    weights = np.exp(-((X - x_query)**2) / (2 * tau**2))
    W = np.diag(weights)
    # Add bias term
    X_b = np.c_[np.ones(len(X)), X]
    query_b = np.array([1, x_query])
    # Closed-form weighted OLS: theta = (X^T W X)^(-1) X^T W y
    theta = np.linalg.inv(X_b.T.dot(W).dot(X_b)).dot(X_b.T).dot(W).dot(y)
    return query_b.dot(theta)

print('Prediction at query point 3.0 (tau=0.5):', lwr_predict(3.0, X, y, tau=0.5))
print('True sine value at 3.0:', np.sin(3.0))
```
