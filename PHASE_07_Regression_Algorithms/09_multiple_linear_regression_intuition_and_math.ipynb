## 📝 Core Concepts & Explanations

### Multiple Linear Regression
Fits a hyperplane in $n$-dimensional space:
$$y = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \dots + \theta_n x_n = X\theta$$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression

diabetes = load_diabetes()
X = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
y = diabetes.target

mlr = LinearRegression().fit(X, y)
print('Model Intercept:', mlr.intercept_)
print('First 3 Coefficients:', mlr.coef_[:3])
```

### Exercises
1. Train a model using only the 'age', 'sex', and 'bmi' features from the diabetes dataset.

```python
X_sub = X[['age', 'sex', 'bmi']]
mlr_sub = LinearRegression().fit(X_sub, y)
print('Subset coefficients:', mlr_sub.coef_)
```
