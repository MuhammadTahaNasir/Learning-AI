## 📝 Core Concepts & Explanations

### Single Variable Regression
Implementing predictions $\hat{y} = mx + c$ and MSE calculation directly using basic NumPy operations.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
X = np.random.rand(50) * 10
y = 2.5 * X + 1.5 + np.random.randn(50) * 2

# Fit parameters analytically
m = np.cov(X, y)[0, 1] / np.var(X)
c = np.mean(y) - m * np.mean(X)
y_pred = m * X + c
print(f'Parameters: m = {m:.3f}, c = {c:.3f}')
```

### Exercises
1. Calculate the residuals ($y - \hat{y}$) and print their mean (which should be close to 0).

```python
residuals = y - y_pred
print(f'Mean of residuals: {residuals.mean():.6f}')
```
