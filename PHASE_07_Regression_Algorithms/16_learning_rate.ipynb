# 📉 Topic 16: Learning Rate

**Summary:** Influence of learning rate on gradient descent convergence.

---

## 📝 Core Concepts & Explanations

### Hyperparameter: Learning Rate ($\alpha$)
- Too small: very slow convergence.
- Too large: divergence, cost increases, oscillation.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

x_start = 10.0

def run_gd(lr, epochs=5):
    x = x_start
    history = []
    for _ in range(epochs):
        grad = 2*x
        x = x - lr * grad
        history.append(x)
    return history

print('Optimal LR (0.1):', run_gd(0.1))
print('Diverging LR (1.1):', run_gd(1.1))
```
