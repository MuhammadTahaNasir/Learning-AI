## 📝 Core Concepts & Explanations

### Multivariate Vectorized Formulation
Gradient: $\nabla_\theta J = \frac{2}{m} X^T(X\theta - y)$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import StandardScaler

diabetes = load_diabetes()
X = diabetes.data[:, :3] # use first 3 features
y = diabetes.target

# Standardize features to aid GD convergence
X_scaled = StandardScaler().fit_transform(X)
X_b = np.c_[np.ones((len(X_scaled), 1)), X_scaled]

theta = np.zeros(4)
lr = 0.1
m_samples = len(y)

for epoch in range(200):
    gradients = (2/m_samples) * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - lr * gradients
print('Fitted coefficients [Intercept, Beta1, Beta2, Beta3]:', theta)
```

### Exercises
1. Calculate the final objective cost $J(\theta)$ on the training dataset.

```python
cost = (1 / (2*m_samples)) * np.sum((X_b.dot(theta) - y)**2)
print(f'Final cost J: {cost:.4f}')
```
