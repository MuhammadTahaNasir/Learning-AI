## 📝 Core Concepts & Explanations

### ElasticNet Regularization
Combines L1 and L2 penalties using mixing ratio $r$ (l1_ratio):
$$J(\theta) = MSE + r \cdot \alpha \sum |\theta_j| + \frac{1-r}{2} \cdot \alpha \sum \theta_j^2$$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
from sklearn.linear_model import ElasticNet
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
X, y = diabetes.data, diabetes.target

enet = ElasticNet(alpha=0.1, l1_ratio=0.5)
enet.fit(X, y)
print('ElasticNet score:', enet.score(X, y))
```

### Exercises
1. Run ElasticNet with l1_ratio = 1.0 and confirm if it yields identical coefficients to Lasso.

```python
from sklearn.linear_model import Lasso
import numpy as np
enet_lasso = ElasticNet(alpha=0.1, l1_ratio=1.0).fit(X, y)
lasso_comp = Lasso(alpha=0.1).fit(X, y)
print('Difference:', np.max(np.abs(enet_lasso.coef_ - lasso_comp.coef_)))
```
