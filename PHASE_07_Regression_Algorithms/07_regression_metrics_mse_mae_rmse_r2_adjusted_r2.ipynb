## 📝 Core Concepts & Explanations

### Regression Metrics
- **Mean Squared Error (MSE):** $\frac{1}{n} \sum (y_i - \hat{y}_i)^2$
- **Mean Absolute Error (MAE):** $\frac{1}{n} \sum |y_i - \hat{y}_i|$
- **Root Mean Squared Error (RMSE):** $\sqrt{MSE}$
- **R-Squared ($R^2$):** $1 - \frac{SS_{res}}{SS_{tot}}$
- **Adjusted $R^2$:** $1 - (1-R^2)\frac{n-1}{n-p-1}$ (penalizes unnecessary features)

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

y_true = np.array([3.0, -0.5, 2.0, 7.0])
y_pred = np.array([2.5, 0.0, 2.0, 8.0])

mse = mean_squared_error(y_true, y_pred)
mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_true, y_pred)

print(f'MSE: {mse}, MAE: {mae}, RMSE: {rmse}, R2: {r2}')
```

### Exercises
1. Compute Adjusted R2 for a model with $R^2 = 0.85$, $n = 100$ samples, and $p = 5$ features.

```python
r2_val = 0.85
n = 100
p = 5
adj_r2 = 1 - (1 - r2_val) * (n - 1) / (n - p - 1)
print(f'Adjusted R2: {adj_r2:.4f}')
```
