## 📝 Core Concepts & Explanations

### Ridge Regularization
Applies an $L_2$ squared penalty to coefficients to reduce collinearity and prevent overfitting:
$$J(\theta) = MSE + \alpha \sum_{j=1}^n \theta_j^2$$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
X, y = diabetes.data, diabetes.target

ridge = Ridge(alpha=1.0)
ridge.fit(X, y)
print('Ridge Coefficients norm:', np.linalg.norm(ridge.coef_))
```

### Exercises
1. Compare Ridge coefficients norm with standard OLS Linear Regression norm.

```python
from sklearn.linear_model import LinearRegression
ols = LinearRegression().fit(X, y)
print('OLS Coefficients norm:', np.linalg.norm(ols.coef_))
print('Notice that Ridge shrinks the coefficient magnitudes.')
```
