## 📝 Core Concepts & Explanations

### Optimization via Gradient Descent
Minimizing an objective function by updating values in the opposite direction of the gradient:
$$x_{new} := x_{old} - \alpha f'(x_{old})$$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

# Function: f(x) = x^2 - 4x + 4, Derivative: f'(x) = 2x - 4
x = 10.0 # start point
lr = 0.1

for epoch in range(1, 11):
    grad = 2*x - 4
    x = x - lr * grad
    print(f'Epoch {epoch}: x = {x:.5f}, f(x) = {x**2 - 4*x + 4:.5f}')
```
