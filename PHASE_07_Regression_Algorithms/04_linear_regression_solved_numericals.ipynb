## 📝 Core Concepts & Explanations

### Analytical Calculation of Coefficients
For $y = mx + c$:
- $m = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2}$
- $c = \bar{y} - m\bar{x}$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 4, 6])

x_bar, y_bar = np.mean(X), np.mean(y)
m = np.sum((X - x_bar) * (y - y_bar)) / np.sum((X - x_bar)**2)
c = y_bar - m * x_bar
print(f'Manual slope: {m}, intercept: {c}')

lr = LinearRegression().fit(X.reshape(-1, 1), y)
print(f'Sklearn slope: {lr.coef_[0]}, intercept: {lr.intercept_}')
```

### Exercises
1. Calculate $m$ and $c$ manually for inputs $X=[1, 2, 3]$ and $Y=[2, 4, 5]$.

```python
X_ex = np.array([1, 2, 3])
y_ex = np.array([2, 4, 5])
m_ex = np.sum((X_ex - np.mean(X_ex)) * (y_ex - np.mean(y_ex))) / np.sum((X_ex - np.mean(X_ex))**2)
c_ex = np.mean(y_ex) - m_ex * np.mean(X_ex)
print(f'Manual slope: {m_ex:.4f}, intercept: {c_ex:.4f}')
```
