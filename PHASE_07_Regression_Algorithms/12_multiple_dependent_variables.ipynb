## 📝 Core Concepts & Explanations

### Multi-Output Regression
Predicting multiple continuous target values simultaneously: $Y \in \mathbb{R}^{m \times k}$.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.random.rand(100, 3)
# Targets are correlated with features
y1 = X[:, 0] * 2 + X[:, 1]
y2 = X[:, 2] * 3 - X[:, 1]
y = np.column_stack([y1, y2])

multi_lr = LinearRegression().fit(X, y)
print('Coefficient matrix shape:', multi_lr.coef_.shape)
print('Coefficients:\n', multi_lr.coef_)
```

### Exercises
1. Predict targets for a test point $X = [[0.5, 0.5, 0.5]]$.

```python
pred = multi_lr.predict([[0.5, 0.5, 0.5]])
print('Predicted outputs:', pred)
```
