## 📝 Core Concepts & Explanations

### Gradient Descent Variants
- **Batch GD:** Calculates gradient on the *entire* dataset.
- **Stochastic GD (SGD):** Calculates gradient on *one* sample at a time (noisy but fast).
- **Mini-batch GD:** Calculates gradient on a small batch of size $B$ (compromise).

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

X = np.random.rand(100, 1)
y = 3*X + 2 + np.random.randn(100, 1)*0.1
X_b = np.c_[np.ones((100, 1)), X] # add bias

# SGD Implementation
theta = np.random.randn(2, 1)
lr = 0.01
for i in range(100):
    rand_idx = np.random.randint(100)
    xi = X_b[rand_idx:rand_idx+1]
    yi = y[rand_idx:rand_idx+1]
    gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
    theta = theta - lr * gradients
print('SGD final theta:', theta.flatten())
```
