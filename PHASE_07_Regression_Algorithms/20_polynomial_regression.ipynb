## 📝 Core Concepts & Explanations

### Polynomial Mapping
Mapping features $X \to [1, X, X^2, X^3]$ to fit non-linear curved bounds using standard OLS solvers.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Generate curved data
np.random.seed(42)
X = np.sort(np.random.rand(40, 1) * 6, axis=0)
y = 0.5 * X**3 - 3 * X**2 + X + np.random.randn(40, 1) * 2

poly = PolynomialFeatures(degree=3, include_bias=False)
X_poly = poly.fit_transform(X)

model = LinearRegression().fit(X_poly, y)
print('Polynomial Coefficients:', model.coef_)
```

### Exercises
1. Calculate predictions and print the R2 score of the polynomial model.

```python
r2 = model.score(X_poly, y)
print(f'Polynomial R2 score: {r2:.4f}')
```
