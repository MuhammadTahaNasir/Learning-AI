## 📝 Core Concepts & Explanations

### Variance Decomposition
- **SST (Sum of Squares Total):** $\sum (y_i - \bar{y})^2$ (Total variation in target)
- **SSR (Sum of Squares Regression):** $\sum (\hat{y}_i - \bar{y})^2$ (Explained variation)
- **SSE (Sum of Squares Error):** $\sum (y_i - \hat{y}_i)^2$ (Unexplained variation)
Relationship: $SST = SSR + SSE$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

y_true = np.array([10, 12, 15, 18, 20])
y_pred = np.array([10.5, 11.8, 14.2, 18.5, 19.3])
y_bar = np.mean(y_true)

sst = np.sum((y_true - y_bar)**2)
ssr = np.sum((y_pred - y_bar)**2)
sse = np.sum((y_true - y_pred)**2)

print(f'SST: {sst:.3f}, SSR: {ssr:.3f}, SSE: {sse:.3f}')
print(f'SSR + SSE = {ssr+sse:.3f} (SST={sst:.3f})')
```
