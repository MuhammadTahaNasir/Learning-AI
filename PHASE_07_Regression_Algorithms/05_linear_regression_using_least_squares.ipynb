## 📝 Core Concepts & Explanations

### OLS Normal Equation
The analytical closed-form solution to minimize MSE in matrix form is:
$$\theta = (X^T X)^{-1} X^T y$$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import pandas as pd
import seaborn as sns

tips = sns.load_dataset('tips')[['total_bill', 'size', 'tip']].dropna()
X = tips[['total_bill', 'size']].values
# Add intercept column of ones
X_b = np.c_[np.ones((len(X), 1)), X]
y = tips['tip'].values

# Compute normal equation
theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
print('OLS Normal Equation Coefficients [Intercept, total_bill, size]:')
print(theta)
```

### Exercises
1. Predict the tip for total_bill = 25.0 and size = 3 using the computed theta vector.

```python
sample = np.array([1, 25.0, 3])
pred = sample.dot(theta)
print(f'Predicted tip: ${pred:.2f}')
```
