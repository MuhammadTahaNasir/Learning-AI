## 📝 Core Concepts & Explanations

### Simple Linear Regression
Fits a straight line to the data: $y = mx + c$, where $m$ is the slope and $c$ is the intercept.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

tips = sns.load_dataset('tips')
X = tips[['total_bill']]
y = tips['tip']

model = LinearRegression()
model.fit(X, y)
print('Slope (m):', model.coef_[0])
print('Intercept (c):', model.intercept_)
```

### Exercises
1. Use the trained model to predict the tip for a total bill of $30.00.

```python
pred_tip = model.predict(pd.DataFrame([[30.0]], columns=['total_bill']))[0]
print(f'Predicted tip for $30 bill: ${pred_tip:.2f}')
```
