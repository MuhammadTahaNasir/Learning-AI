## 📝 Core Concepts & Explanations

### Regression Analysis
Regression is a supervised learning task where the goal is to predict a continuous numerical value $Y$ based on one or more input features $X$.
- **Independent Variable ($X$):** The predictor, input, or feature.
- **Dependent Variable ($Y$):** The target, output, or response.
Examples: House size ($X$) vs price ($Y$), study hours ($X$) vs grade ($Y$).
