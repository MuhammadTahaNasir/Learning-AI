## 📝 Core Concepts & Explanations

### Matrix Normal Equation Solver
Solving $(X^T X)^{-1} X^T y$ manually on a small $3 \times 2$ feature dataset.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

# Features: Col 0 is intercept (ones), Col 1 is X1, Col 2 is X2
X = np.array([[1, 2, 3],
              [1, 3, 5],
              [1, 4, 6]])
y = np.array([5, 8, 10])

# Compute beta = (X^T * X)^(-1) * X^T * y
XTX = X.T.dot(X)
beta = np.linalg.inv(XTX).dot(X.T).dot(y)
print('Calculated weights beta:', beta)
```
