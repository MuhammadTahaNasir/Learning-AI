**Summary:** Linear regression assumptions: Linearity, Independence, Homoscedasticity, Normality.

---

## 📝 Core Concepts & Explanations

### Core Assumptions of OLS
1. **Linearity:** Relationship between $X$ and $Y$ is linear.
2. **Independence:** Residuals are independent.
3. **Homoscedasticity:** Constant variance of residuals across predictions.
4. **Normality of Residuals:** Error terms are normally distributed.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate simulated residuals
np.random.seed(42)
residuals = np.random.normal(0, 1, 100)

# Q-Q Plot to check normality assumption
stats.probplot(residuals, dist='norm', plot=plt)
plt.title('Residual Q-Q Plot')
plt.show()
```
