## 📝 Core Concepts & Explanations

### Mathematical Formulation
- **Hypothesis:** $h_\theta(x) = \sum_{j=0}^n \theta_j x_j = \theta^T x$ (where $x_0 = 1$)
- **Cost Function (Mean Squared Error):** $J(\theta) = \frac{1}{2m} \sum_{i=1}^m (h_\theta(x^{(i)}) - y^{(i)})^2$
- **Gradient Descent Update Rule:** \theta_j := \theta_j - \alpha \frac{\partial J(\theta)}{\partial \theta_j}$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate a quadratic cost function curve J(theta)
theta = np.linspace(-10, 10, 100)
J = 0.5 * (theta - 2)**2 + 3

plt.plot(theta, J, label='Cost Function J(theta)')
plt.axvline(2, color='red', linestyle='--', label='Optimal theta=2')
plt.xlabel('theta')
plt.ylabel('J(theta)')
plt.legend()
plt.show()
```

### Exercises
1. Plot the gradient tangent line at $\theta = -5$ for the simulated cost function.

```python
theta_val = -5
J_val = 0.5 * (theta_val - 2)**2 + 3
slope = theta_val - 2
tangent = slope * (theta - theta_val) + J_val
plt.plot(theta, J, label='Cost Function')
plt.plot(theta, tangent, '--', label='Tangent at theta=-5')
plt.scatter([theta_val], [J_val], color='red')
plt.ylim(0, 50)
plt.legend()
plt.show()
```
