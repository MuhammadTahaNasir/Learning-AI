## 📝 Core Concepts & Explanations

### Loop-based Gradient Descent
Updating weights $m$ and $c$ using explicit iterations over data samples instead of vector operations.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np

X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

m, c = 0.0, 0.0
lr = 0.01
n = len(X)

for epoch in range(100):
    grad_m = 0.0
    grad_c = 0.0
    for i in range(n):
        y_pred = m * X[i] + c
        grad_m += (y_pred - y[i]) * X[i]
        grad_c += (y_pred - y[i])
    m = m - lr * (2/n) * grad_m
    c = c - lr * (2/n) * grad_c

print(f'Fitted parameters: m = {m:.3f}, c = {c:.3f}')
```

### Exercises
1. Calculate the final Mean Squared Error of this model.

```python
final_preds = m * X + c
mse = np.mean((y - final_preds)**2)
print(f'Final MSE: {mse:.4f}')
```
