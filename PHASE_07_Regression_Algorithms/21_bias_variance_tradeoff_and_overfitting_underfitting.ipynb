## 📝 Core Concepts & Explanations

### Bias vs Variance Tradeoff
- **High Bias:** Underfitting (model is too simple to capture patterns, eg. degree 1 on curved data).
- **High Variance:** Overfitting (model fits training noise too closely, eg. degree 15 on small dataset).

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

np.random.seed(42)
X = np.random.rand(30, 1) * 4
y = np.sin(X).ravel() + np.random.randn(30) * 0.2

def evaluate_deg(deg):
    poly = PolynomialFeatures(degree=deg)
    X_p = poly.fit_transform(X)
    model = LinearRegression().fit(X_p, y)
    train_mse = mean_squared_error(y, model.predict(X_p))
    return train_mse

print('Underfit Degree 1 Train MSE:', evaluate_deg(1))
print('Overfit Degree 12 Train MSE:', evaluate_deg(12))
```
