## 📝 Core Concepts & Explanations

### Lasso Regularization
Applies an $L_1$ absolute penalty to coefficients, which encourages sparsity (zeroing out weights):
$$J(\theta) = MSE + \alpha \sum_{j=1}^n |\theta_j|$$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
from sklearn.linear_model import Lasso
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
X, y = diabetes.data, diabetes.target

lasso = Lasso(alpha=0.1)
lasso.fit(X, y)
print('Lasso Coefficients:', lasso.coef_)
print('Number of features zeroed out:', np.sum(lasso.coef_ == 0))
```

### Exercises
1. Train Lasso with a very large alpha (e.g., 5.0) and see how many features remain non-zero.

```python
lasso_large = Lasso(alpha=5.0).fit(X, y)
print('Non-zero coefficients count:', np.sum(lasso_large.coef_ != 0))
```
