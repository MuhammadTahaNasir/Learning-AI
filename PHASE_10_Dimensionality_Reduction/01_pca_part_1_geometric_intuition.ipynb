## 📝 Core Concepts & Explanations

### Principal Component Analysis (PCA) Intuition
PCA seeks to find a linear projection of high-dimensional data into a lower-dimensional subspace such that:
1. **Variance is Maximized**: The projected data retains as much spread (information) as possible.
2. **Projection Error is Minimized**: The orthogonal distance from the original points to the projection subspace is minimized.
These two objectives are mathematically equivalent.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# 1. Load real data (Iris) and pick 2 features for 2D visualization
X = load_iris().data[:, :2] # Sepal length and width
X_scaled = StandardScaler().fit_transform(X)

# 2. Apply PCA to find the 1st principal component
pca = PCA(n_components=1)
pca.fit(X_scaled)

# 3. Plot original data and the principal component vector
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], alpha=0.5, label='Standardized Data')

# Draw the principal component axis
mean = np.mean(X_scaled, axis=0)
vector = pca.components_[0] * 3  # Scale for visibility
plt.plot([mean[0] - vector[0], mean[0] + vector[0]], 
         [mean[1] - vector[1], mean[1] + vector[1]], 
         color='red', linewidth=3, label='1st Principal Component (Max Variance)')

plt.title('PCA Geometric Intuition (Iris Dataset)')
plt.xlabel('Sepal Length (Standardized)')
plt.ylabel('Sepal Width (Standardized)')
plt.axis('equal')
plt.grid(True, linestyle='--')
plt.legend()
plt.show()
```
