## 📝 Core Concepts & Explanations

### PCA vs LDA
- **PCA:** Unsupervised. Finds axes that maximize total variance, ignoring class labels.
- **LDA (Linear Discriminant Analysis):** Supervised. Finds axes that maximize the separation (distance) between multiple classes while minimizing the variance within each class.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.preprocessing import StandardScaler

# Load 13-feature Wine dataset
X, y = load_wine(return_X_y=True)
X_scaled = StandardScaler().fit_transform(X)

# Apply Unsupervised PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Apply Supervised LDA
lda = LDA(n_components=2)
X_lda = lda.fit_transform(X_scaled, y)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

scatter1 = ax[0].scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='Set1', edgecolors='k')
ax[0].set_title('PCA Projection (Unsupervised)')
ax[0].set_xlabel('Principal Component 1')
ax[0].set_ylabel('Principal Component 2')

scatter2 = ax[1].scatter(X_lda[:, 0], X_lda[:, 1], c=y, cmap='Set1', edgecolors='k')
ax[1].set_title('LDA Projection (Supervised)')
ax[1].set_xlabel('Linear Discriminant 1')
ax[1].set_ylabel('Linear Discriminant 2')

plt.suptitle('PCA vs LDA on Wine Dataset', fontsize=14, fontweight='bold')
plt.show()
```
