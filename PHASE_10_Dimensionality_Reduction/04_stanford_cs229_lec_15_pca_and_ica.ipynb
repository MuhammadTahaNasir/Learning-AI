## 📝 Core Concepts & Explanations

### PCA vs ICA
- **PCA (Principal Component Analysis):** Finds orthogonal directions of maximum variance. Assumes features are Gaussian-distributed.
- **ICA (Independent Component Analysis):** Finds a non-orthogonal linear transformation that minimizes statistical dependence between components. Assumes features are **Non-Gaussian**. Used extensively to solve the *Blind Source Separation (Cocktail Party) Problem*.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from sklearn.decomposition import FastICA, PCA

# Generate 2 Independent Sources (Sine and Sawtooth)
np.random.seed(42)
time = np.linspace(0, 8, 2000)
s1 = np.sin(2 * time)  # Signal 1
s2 = signal.sawtooth(2 * np.pi * time)  # Signal 2
S = np.c_[s1, s2]

# Mix Data
A = np.array([[1, 1], [0.5, 2]])  # Mixing matrix
X = np.dot(S, A.T)  # Mixed signals observation

# Apply PCA and ICA
pca = PCA(n_components=2)
H_pca = pca.fit_transform(X)  # PCA fails to separate independent non-gaussian sources

ica = FastICA(n_components=2, random_state=42, whiten='unit-variance')
H_ica = ica.fit_transform(X)  # ICA isolates the independent sources

models = [X, S, H_pca, H_ica]
names = ['Mixed Observations', 'True Sources', 'PCA Recovered', 'ICA Recovered']
colors = ['red', 'steelblue']

fig, axes = plt.subplots(4, 1, figsize=(8, 8), sharex=True)
for i, (model, name) in enumerate(zip(models, names)):
    axes[i].set_title(name)
    for sig, color in zip(model.T, colors):
        axes[i].plot(sig, color=color)

plt.tight_layout()
plt.show()
```
