## 📝 Core Concepts & Explanations

### The Mathematics of PCA (5 Steps)
1. **Standardize the dataset**: $Z = \frac{X - \mu}{\sigma}$
2. **Compute Covariance Matrix**: $\Sigma = \frac{1}{n-1} Z^T Z$
3. **Compute Eigendecomposition**: $\Sigma v = \lambda v$ (where $\lambda$ are eigenvalues, $v$ are eigenvectors).
4. **Sort Eigenpairs**: Sort eigenvectors by descending eigenvalues to rank them by explained variance.
5. **Projection Matrix**: Select the top $k$ eigenvectors to form matrix $W$, and project data: $X_{pca} = Z W$.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler

# 1. Load and Standardize Wine Dataset (13 features)
X, y = load_wine(return_X_y=True)
Z = StandardScaler().fit_transform(X)

# 2. Compute Covariance Matrix
cov_matrix = np.cov(Z.T)

# 3 & 4. Eigendecomposition and Sorting
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
sorted_idx = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_idx]

# Visualizations
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# Heatmap of Covariance Matrix
sns.heatmap(cov_matrix, cmap='coolwarm', ax=ax[0], square=True)
ax[0].set_title('Feature Covariance Matrix Heatmap')

# Scree Plot of Eigenvalues
ax[1].plot(range(1, 14), sorted_eigenvalues, 'bo-', linewidth=2)
ax[1].set_title('Scree Plot (Eigenvalues)')
ax[1].set_xlabel('Principal Component Rank')
ax[1].set_ylabel('Eigenvalue (Magnitude of Variance)')
ax[1].grid(True, linestyle=':')
plt.tight_layout()
plt.show()
```
