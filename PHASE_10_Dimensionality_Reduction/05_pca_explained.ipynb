## 📝 Core Concepts & Explanations

### Limitations of Linear PCA
PCA is strictly a **linear** transformation. If the data lies on a complex, non-linear manifold (like a sphere, spiral, or nested rings), standard PCA cannot unroll or unfold the data. It will simply crush the data onto a flat hyperplane.
In such cases, non-linear techniques like **Kernel PCA**, **t-SNE**, or **UMAP** are required.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.decomposition import PCA, KernelPCA

# Generate Non-Linear data (Moons)
X, y = make_moons(n_samples=200, noise=0.05, random_state=42)

# Standard PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Kernel PCA (RBF)
kpca = KernelPCA(n_components=2, kernel='rbf', gamma=15)
X_kpca = kpca.fit_transform(X)

fig, ax = plt.subplots(1, 3, figsize=(15, 4))

ax[0].scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolors='k')
ax[0].set_title('Original Non-Linear Data')

ax[1].scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolors='k')
ax[1].set_title('Standard PCA (Linear Failure)')

ax[2].scatter(X_kpca[:, 0], X_kpca[:, 1], c=y, cmap='viridis', edgecolors='k')
ax[2].set_title('Kernel PCA (RBF Unfolding)')

plt.show()
```
