## 📝 Core Concepts & Explanations

### Ensemble Learning
Ensemble learning relies on the **Wisdom of the Crowd**. By aggregating the predictions of a group of diverse, moderately performing predictors (weak learners), we can often construct a single strong learner.
For this to work effectively, the models should be as **independent** as possible (making different types of errors).

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulating the Law of Large Numbers for Ensembles
coin_tosses = (np.random.rand(10000, 10) < 0.51).astype(int) # 51% biased coin
cumulative_heads_ratio = np.cumsum(coin_tosses, axis=0) / np.arange(1, 10001).reshape(-1, 1)

plt.figure(figsize=(8, 4))
plt.plot(cumulative_heads_ratio[:, 0], label='1 Weak Learner (Coin)')
plt.plot(cumulative_heads_ratio[:, 1:4], alpha=0.5)
plt.axhline(y=0.51, color='k', linestyle='--', label='51% Probability')
plt.axhline(y=0.50, color='r', linestyle='--', label='50% Boundary')
plt.title('Law of Large Numbers in Ensembles')
plt.xlabel('Number of Tosses (or Trees)')
plt.ylabel('Heads Ratio')
plt.legend()
plt.show()
```
