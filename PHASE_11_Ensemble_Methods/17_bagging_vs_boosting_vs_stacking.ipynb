## 📝 Core Concepts & Explanations

### The Big Three Summarized
| Feature | Bagging | Boosting | Stacking |
|---|---|---|---|
| Paradigm | Parallel | Sequential | Meta-Learning |
| Focus | Reduce Variance | Reduce Bias | Learn best combinations |
| Base Models | Same type (homogeneous) | Same type (homogeneous) | Different types (heterogeneous) |
| Algorithm | Random Forest | XGBoost | StackingClassifier |
