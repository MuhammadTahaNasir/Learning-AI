## 📝 Core Concepts & Explanations

### Bagging vs Boosting
- **Bagging:** Parallel. Each model is built independently. Fixes Variance. Prevents Overfitting. Use complex trees.
- **Boosting:** Sequential. Each model corrects the previous model's errors. Fixes Bias. Prone to Overfitting. Use simple stumps.
