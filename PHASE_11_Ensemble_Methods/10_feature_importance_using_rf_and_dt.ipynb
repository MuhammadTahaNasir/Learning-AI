## 📝 Core Concepts & Explanations

### Feature Importance
Random Forests measure how much a feature reduces impurity (Gini/Entropy) across all nodes in all trees. It is a fast, built-in way to perform Feature Selection.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier

data = load_breast_cancer()
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(data.data, data.target)

importances = rf.feature_importances_
indices = np.argsort(importances)[-8:] # Top 8 features

plt.figure(figsize=(8, 4))
plt.barh(range(len(indices)), importances[indices], color='darkorange')
plt.yticks(range(len(indices)), [data.feature_names[i] for i in indices])
plt.title('Random Forest Feature Importances')
plt.show()
```
