## 📝 Core Concepts & Explanations

### Exploring the Forest
A Random Forest is just an array of `DecisionTreeClassifier` objects. You can access individual trees using the `.estimators_` attribute.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)
rf = RandomForestClassifier(n_estimators=3, max_depth=2, random_state=42)
rf.fit(X, y)

# Accessing the 1st Tree in the Forest
first_tree = rf.estimators_[0]
print(f'Type of 1st tree: {type(first_tree)}')
print(f'Depth of 1st tree: {first_tree.get_depth()}')
```

### Exercises
1. Since you can access individual trees using `.estimators_`, could you plot or print the rules of a single tree inside a Random Forest?

```python
from sklearn.tree import export_text
from sklearn.datasets import load_iris

# Yes! We can print the text representation of the 1st tree.
tree_rules = export_text(first_tree, feature_names=load_iris().feature_names)
print(tree_rules)
```
