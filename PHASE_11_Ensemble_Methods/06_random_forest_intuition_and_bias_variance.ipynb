## 📝 Core Concepts & Explanations

### Random Forest Intuition
A Random Forest is fundamentally a Bagging ensemble of Decision Trees. However, it introduces an extra layer of randomness: when splitting a node, it only considers a **random subset of features** (usually $\sqrt{n}$).
This forces the trees to be even more uncorrelated, which drastically reduces Variance at the cost of a slight increase in Bias.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split

X, y = load_wine(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

rf_clf = RandomForestClassifier(n_estimators=100, max_leaf_nodes=16, random_state=42)
rf_clf.fit(X_train, y_train)

print(f'Random Forest Accuracy on Wine: {rf_clf.score(X_test, y_test):.3f}')
```
