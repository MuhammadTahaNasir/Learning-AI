## 📝 Core Concepts & Explanations

### Voting Regressor
In regression, voting implies averaging the numerical predictions of all the base estimators.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import VotingRegressor, RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

data = fetch_california_housing()
X_train, X_test, y_train, y_test = train_test_split(data.data[:1000], data.target[:1000], random_state=42)

reg1 = LinearRegression()
reg2 = RandomForestRegressor(random_state=42, n_estimators=50)
reg3 = DecisionTreeRegressor(random_state=42, max_depth=5)

ereg = VotingRegressor(estimators=[('lr', reg1), ('rf', reg2), ('dt', reg3)])
ereg.fit(X_train, y_train)
reg1.fit(X_train, y_train) # Fit standalone reg1 so we can compare its prediction

print(f'Linear Reg MSE: {mean_squared_error(y_test, reg1.predict(X_test)):.3f}')
print(f'Voting Reg MSE: {mean_squared_error(y_test, ereg.predict(X_test)):.3f}')
```
