## 📝 Core Concepts & Explanations

### Power Transforms
- **Box-Cox**: Requires strictly positive data ($x > 0$). Uses parameter estimation to select the optimal power transform.
- **Yeo-Johnson**: Generalization that supports zero and negative values.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
from sklearn.preprocessing import PowerTransformer

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)[['Fare']].dropna()

pt = PowerTransformer(method='yeo-johnson')
df['Yeo_Fare'] = pt.fit_transform(df[['Fare']])
print('Yeo-Johnson Transformed Skewness:', df['Yeo_Fare'].skew())
```

### Exercises
1. Fit the Box-Cox transformer to the 'Fare' column (filter out values $\leq 0$ first) and print its optimal lambda parameter.

```python
positive_fares = df[df['Fare'] > 0][['Fare']]
bc = PowerTransformer(method='box-cox')
bc.fit(positive_fares)
print('Optimal Box-Cox lambda:', bc.lambdas_)
```
