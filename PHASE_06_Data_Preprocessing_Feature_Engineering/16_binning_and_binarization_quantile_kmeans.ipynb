## 📝 Core Concepts & Explanations

### Discretization & Binarization
- **Binning**: Converting a continuous feature into discrete bins (e.g. groups of ages). Types: Equal-width, Quantile (equal-frequency), or K-Means.
- **Binarization**: Thresholding features to binary $[0, 1]$ flags.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
from sklearn.preprocessing import KBinsDiscretizer, Binarizer

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)[['Age', 'Fare']].dropna()

# Binning Age into 5 bins
kbd = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='quantile')
df['Age_Binned'] = kbd.fit_transform(df[['Age']])

# Binarizing Fare at 20.0
binarizer = Binarizer(threshold=20.0)
df['Fare_Over_20'] = binarizer.transform(df[['Fare']])
print(df.head())
```
