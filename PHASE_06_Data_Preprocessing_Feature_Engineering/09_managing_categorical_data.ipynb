## 📝 Core Concepts & Explanations

### Categorical Variables
- **Nominal**: Variables with no implicit ordering (e.g., City, Gender).
- **Ordinal**: Variables with a defined, sequential order (e.g., Education: High School < Bachelor's < Ph.D.).
- High cardinality refers to categorical features with many unique levels, which can expand dimensions significantly if OHE is applied.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

print('=== Unique Value Counts (Cardinality) ===')
print(df.select_dtypes(include=['object']).nunique())
```
