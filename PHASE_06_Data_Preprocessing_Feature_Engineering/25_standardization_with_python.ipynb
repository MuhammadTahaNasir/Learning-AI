## 📝 Core Concepts & Explanations

### StandardScaler Code Verification
Scaling values to zero mean and unit variance.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)[['Age']].dropna()

scaler = StandardScaler()
df['Standardized_Age'] = scaler.fit_transform(df[['Age']])
print(df.describe())
```

### Exercises
1. Standardize the 'Fare' column and check if the mean is approximately 0.

```python
fare_col = pd.read_csv(url)[['Fare']].dropna()
scaled_fare = StandardScaler().fit_transform(fare_col)
print('Scaled mean:', scaled_fare.mean().round(6))
```
