## 📝 Core Concepts & Explanations

### The Curse of Dimensionality
As the number of features (dimensions) increases, the volume of the space grows exponentially. This causes:
1. Sparsity of training samples.
2. Distances between points to converge (making clustering and distance-based models less effective).

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import numpy as np
from scipy.spatial.distance import pdist

# Generate random points in 2D vs 100D
pts_2d = np.random.rand(100, 2)
pts_100d = np.random.rand(100, 100)

dist_2d = pdist(pts_2d)
dist_100d = pdist(pts_100d)

print('Ratio of (Max - Min) / Min distance:')
print('2D:', (dist_2d.max() - dist_2d.min()) / dist_2d.min())
print('100D:', (dist_100d.max() - dist_100d.min()) / dist_100d.min())
```
