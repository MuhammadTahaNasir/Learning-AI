## 📝 Core Concepts & Explanations

### Label vs Ordinal Encoding
- `LabelEncoder`: Encodes values to 0..n-1 integers. Typically used for target variables.
- `OrdinalEncoder`: Encodes features. Allows custom ordering mapping for categories.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

df = pd.DataFrame({'Size': ['Small', 'Medium', 'Large', 'Medium']})

# Explicitly mapping order
encoder = OrdinalEncoder(categories=[['Small', 'Medium', 'Large']])
df['Size_Encoded'] = encoder.fit_transform(df[['Size']])
print(df)
```
