## 📝 Core Concepts & Explanations

### Missing Indicators
Sometimes the *absence* of data is itself a strong predictor. We create binary columns where $1$ denotes missing and $0$ denotes present before imputing.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
from sklearn.impute import MissingIndicator

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)[['Age']]

indicator = MissingIndicator()
missing_mask = indicator.fit_transform(df)
df['Age_Was_Missing'] = missing_mask.astype(int)
print(df.head(10))
```

### Exercises

```python
from sklearn.impute import SimpleImputer
imputer_with_indicator = SimpleImputer(strategy='mean', add_indicator=True)
res = imputer_with_indicator.fit_transform(df[['Age']])
print('Output shape after imputation + indicator:', res.shape)
```
