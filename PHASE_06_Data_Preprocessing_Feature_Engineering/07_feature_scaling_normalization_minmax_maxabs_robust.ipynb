## 📝 Core Concepts & Explanations

### Range Scalers
- **MinMaxScaler**: Scales to a range (usually $[0, 1]$): $x_{scaled} = \frac{x - x_{min}}{x_{max} - x_{min}}$.
- **MaxAbsScaler**: Scales to $[-1, 1]$ by dividing by max absolute value: $x_{scaled} = \frac{x}{|x|_{max}}$.
- **RobustScaler**: Scales using median and IQR, making it robust to outliers: $x_{scaled} = \frac{x - Median}{IQR}$.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, RobustScaler

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)[['Age', 'Fare']].dropna()

minmax = MinMaxScaler()
robust = RobustScaler()

df_minmax = pd.DataFrame(minmax.fit_transform(df), columns=df.columns)
df_robust = pd.DataFrame(robust.fit_transform(df), columns=df.columns)
print('MinMax range of Fare:', df_minmax['Fare'].min(), 'to', df_minmax['Fare'].max())
print('Robust range of Fare:', df_robust['Fare'].min(), 'to', df_robust['Fare'].max())
```
