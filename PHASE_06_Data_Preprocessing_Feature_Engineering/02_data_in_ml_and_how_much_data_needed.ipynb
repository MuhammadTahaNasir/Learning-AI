## 📝 Core Concepts & Explanations

### How Much Data is Needed?
Common guidelines include:
- **Rule of 10**: At least 10 times as many training examples as there are features.
- **Complexity Scaling**: High-bias models (e.g., Naive Bayes) need less data; high-variance models (e.g., Deep Learning) need orders of magnitude more.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
import numpy as np

# Calculate feature-to-sample ratio
features_count = 10
recommended_samples = features_count * 10
print(f'For {features_count} features, we need at least {recommended_samples} samples.')
```
