## 📝 Core Concepts & Explanations

### MinMaxScaler Code Verification
Normalizing values to a strict $[0, 1]$ range.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)[['Fare']].dropna()

scaler = MinMaxScaler()
df['Normalized_Fare'] = scaler.fit_transform(df[['Fare']])
print(df.describe())
```

### Exercises
1. Normalize the 'Age' column of the Titanic dataset and print the minimum and maximum scaled values.

```python
age_col = pd.read_csv(url)[['Age']].dropna()
scaled_age = MinMaxScaler().fit_transform(age_col)
print('Min scaled age:', scaled_age.min())
print('Max scaled age:', scaled_age.max())
```
