## 📝 Core Concepts & Explanations

### Binarizer Code Verification
Thresholding features to binary variables.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
from sklearn.preprocessing import Binarizer

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)[['Fare']].dropna()

binarizer = Binarizer(threshold=50.0)
df['High_Fare'] = binarizer.transform(df[['Fare']])
print(df['High_Fare'].value_counts())
```

### Exercises
1. Binarize the 'Age' column of the Titanic dataset at threshold 60.0 to identify elderly passengers.

```python
age_col = pd.read_csv(url)[['Age']].dropna()
is_elderly = Binarizer(threshold=60.0).transform(age_col)
print('\nElderly counts:\n', pd.Series(is_elderly.flatten()).value_counts())
```
