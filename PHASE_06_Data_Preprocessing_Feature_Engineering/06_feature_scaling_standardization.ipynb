## 📝 Core Concepts & Explanations

### Standardization
Standardization transforms features to have a mean of 0 and a standard deviation of 1:
$$z = \frac{x - \mu}{\sigma}$$

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)[['Age', 'Fare']].dropna()

scaler = StandardScaler()
scaled_features = scaler.fit_transform(df)
df_scaled = pd.DataFrame(scaled_features, columns=df.columns)
print('Scaled Means:\n', df_scaled.mean().round(4))
print('\nScaled Std Devs:\n', df_scaled.std().round(4))
```
