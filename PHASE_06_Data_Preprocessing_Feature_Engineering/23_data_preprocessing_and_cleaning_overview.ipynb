## 📝 Core Concepts & Explanations

### Preprocessing Summary Checklist
1. Resolve Nulls (Imputation or Indicators).
2. Encode Categories (One-Hot or Ordinal).
3. Scale numeric parameters (Standardization or MinMaxScaler).
4. Correct skewed variables (Power Transformations).
5. Apply pipeline to train and test splits to prevent leakage.
