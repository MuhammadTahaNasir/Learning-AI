## 📝 Core Concepts & Explanations

### Outlier Methods
- **Z-Score Method**: Points beyond $\pm 3$ standard deviations from the mean (safe for Gaussian data).
- **IQR Method**: Bounds are $Q_1 - 1.5 \times IQR$ and $Q_3 + 1.5 \times IQR$.
- **Winsorization**: Capping outlier values at specific percentiles (e.g., 1st and 99th).

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)[['Fare']].dropna()

# IQR Calculation
q1 = df['Fare'].quantile(0.25)
q3 = df['Fare'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = df[(df['Fare'] < lower_bound) | (df['Fare'] > upper_bound)]
print(f'Found {len(outliers)} outliers using IQR method.')
```
