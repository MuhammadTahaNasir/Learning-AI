## 📝 Core Concepts & Explanations

### ColumnTransformer & Pipelines
We construct pipelines to chain imputations, encodings, and scaling steps dynamically across heterogeneous column groupings, ensuring a clean and reproducible pipeline.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)[['Age', 'Fare', 'Sex', 'Embarked', 'Pclass']].dropna()
# Drop Pclass from columns so we can use same num/cat features

num_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scale', StandardScaler())
])

cat_pipeline = Pipeline([
    ('encode', OneHotEncoder(drop='first', sparse_output=False))
])

preprocessor = ColumnTransformer([
    ('num', num_pipeline, ['Age', 'Fare']),
    ('cat', cat_pipeline, ['Sex', 'Embarked'])
])

processed_data = preprocessor.fit_transform(df)
print('Processed output shape:', processed_data.shape)
```
