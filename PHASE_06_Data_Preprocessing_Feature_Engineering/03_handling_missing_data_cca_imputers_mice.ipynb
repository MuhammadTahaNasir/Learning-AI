## 📝 Core Concepts & Explanations

### Missing Data Mechanisms & Treatments
- **MCAR** (Missing Completely at Random), **MAR** (Missing at Random), **MNAR** (Missing Not at Random).
- **CCA** (Trimming missing rows): Only safe if MCAR and missingness is <5%.
- **Simple Imputation**: Mean/Median/Mode substitution.
- **MICE** (Multivariate Imputation by Chained Equations): Fills missing values using regression models on other variables iteratively.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df_missing = pd.read_csv(url)[['Age', 'Fare']]

# Simple Imputer
si = SimpleImputer(strategy='median')
df_si = pd.DataFrame(si.fit_transform(df_missing), columns=df_missing.columns)

# MICE (Iterative Imputer)
mice = IterativeImputer(max_iter=10, random_state=42)
df_mice = pd.DataFrame(mice.fit_transform(df_missing), columns=df_missing.columns)
print('Original Nulls:', df_missing['Age'].isnull().sum())
print('Simple Imputed Nulls:', df_si['Age'].isnull().sum())
print('MICE Imputed Nulls:', df_mice['Age'].isnull().sum())
```
