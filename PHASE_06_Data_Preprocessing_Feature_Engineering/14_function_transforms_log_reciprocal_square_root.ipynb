## 📝 Core Concepts & Explanations

### Function Transformations
Right-skewed data can be normalized using mathematical transformations:
- **Log Transform**: $y = \log(x + 1)$. Essential for variables with exponential distribution.
- **Reciprocal Transform**: $y = 1 / x$.
- **Square Root**: $y = \sqrt{x}$.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import FunctionTransformer

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)[['Fare']].dropna()

log_transformer = FunctionTransformer(np.log1p)
df['Log_Fare'] = log_transformer.transform(df[['Fare']])
print('Original skewness:', df['Fare'].skew())
print('Log Transformed skewness:', df['Log_Fare'].skew())
```
