## 📝 Core Concepts & Explanations

### Data Science Lifecycle & Preprocessing
The data science lifecycle is typically modeled using frameworks like **CRISP-DM** (Cross-Industry Standard Process for Data Mining) or **OSEMN**:
1. **O**btain, 2. **S**crub, 3. **E**xplore, 4. **M**odel, 5. i**N**terpret.
Data Preprocessing sits firmly in the **Scrub** phase and is estimated to take up to **70-80%** of a data scientist's time.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd

# Load Titanic dataset to represent real-world data
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
print('=== Basic Scrubbing Checklist ===')
print('Dataset Shape:', df.shape)
print('\nMissing values:\n', df.isnull().sum())
```
