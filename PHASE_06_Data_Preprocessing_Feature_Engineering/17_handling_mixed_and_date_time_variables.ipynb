## 📝 Core Concepts & Explanations

### Mixed Data Processing
- **Mixed variables**: Features containing both numbers and text (e.g., ticket codes). We split these into separate numeric and categorical columns.
- **DateTime variables**: Features representing timestamps. We extract year, month, day, day of week, or hour.

## 💻 Code Demonstration
Run the cells below to see the concepts in action:

```python
import pandas as pd

# Alphanumeric splitting
df = pd.DataFrame({'Ticket': ['A/5 21171', 'PC 17599', '3101298']})
df['Ticket_Num'] = df['Ticket'].str.extract(r'(\d+)')
df['Ticket_Cat'] = df['Ticket'].str.extract(r'([a-zA-Z/\.]+)')
df['Ticket_Cat'] = df['Ticket_Cat'].fillna('Numeric')
print(df)
```

1. Parse the string date series `['2025-08-30 14:30:00', '2025-08-31 15:45:00']` and extract the Hour and Day of Week features.

```python
dates = pd.Series(pd.to_datetime(['2025-08-30 14:30:00', '2025-08-31 15:45:00']))
print('Hours:', dates.dt.hour.values)
print('Days of week:', dates.dt.dayofweek.values)
```
